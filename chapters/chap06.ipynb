# Вернуть значения

В предыдущих главах мы использовали встроенные функции-например, `abs` и `round`-и функции в математическом модуле-например, `sqrt` и `pow`.
Когда вы называете одну из этих функций, он возвращает значение, которое вы можете присвоить переменной или использовать как часть выражения.

Функции, которые мы написали до сих пор, разные.
Некоторые используют функцию `print` для отображения значений, а некоторые используют функции черепахи для рисования фигур.
Но они не возвращают значения, которые мы назначаем переменным или используем в выражениях.

В этой главе мы посмотрим, как писать функции, которые возвращают значения.

## Некоторые функции имеют возвратные значения

Когда вы называете функцию, такую как `math.sqrt`, результат называется ** возвращающим значением **.
Если вызов функции появляется в конце ячейки, Юпитер немедленно отображает возвратное значение.

```python
import math

math.sqrt(42 / math.pi)
```

Если вы назначаете возвратное значение переменной, оно не отображается.

```python
radius = math.sqrt(42 / math.pi)
```

Но вы можете отобразить его позже.

```python
radius
```

Или вы можете использовать возвратное значение как часть выражения.

```python
radius + math.sqrt(42 / math.pi)
```

Вот пример функции, которая возвращает значение.

```python
def circle_area(radius):
    area = math.pi * radius**2
    return area
```

`circle_area` принимает `radius` как параметр и вычисляет область круга с этим радиусом.

Последняя строка - это оператор `return`, который возвращает значение `area`.

Если мы называем функцию такой, Jupyter отображает возвратное значение.

```python
circle_area(radius)
```

Мы можем назначить возвращаемое значение переменной.

```python
a = circle_area(radius)
```

Или используйте его как часть выражения.

```python
circle_area(radius) + 2 * circle_area(radius / 2)
```

Позже мы можем отобразить значение переменной, которой мы присвоили результат.

```python
a
```

Но мы не можем получить доступ `area`.

```python
%%expect NameError

area
```

`area` является локальной переменной в функции, поэтому мы не можем получить к ней доступ извне функции.

## возвращаемые значения и условия

Если Python не предоставил `abs`, мы могли бы написать это так.

```python
def absolute_value(x):
    if x < 0:
        return -x
    else:
        return x
```

Если `x` отрицательный, первый оператор `return` возвращает `-x`, а функция заканчивается немедленно.
В противном случае, второй оператор `return` возвращает `x`, а функция заканчивается.
Так что эта функция верна.

Однако, если вы поместите операторы `return` в условные, вы должны убедиться, что каждый возможный путь через программу достигает оператора `return`.
Например, вот неправильная версия `absolute_value`.

```python
def absolute_value_wrong(x):
    if x < 0:
        return -x
    if x > 0:
        return x
```

Вот что произойдет, если мы назваем эту функцию `0` в качестве аргумента.

```python
absolute_value_wrong(0)
```

Мы ничего не получаем!Вот проблема: когда `x`] является `0`, ни одно условие не является истинной, и функция заканчивается без удара `return`] оператора, что означает, что возвратное значение `None`, поэтому Юпитер ничего не отображает.

В качестве другого примера, вот версия `absolute_value` с дополнительным `return`] в конце.

```python
def absolute_value_extra_return(x):
    if x < 0:
        return -x
    else:
        return x
    
    return 'This is dead code'
```

Если `x` отрицательный, первый оператор `return` выполняется, а функция заканчивается.
В противном случае второй оператор `return` работает, и функция заканчивается.
В любом случае, мы никогда не доберемся до третьего оператора `return` - так что он никогда не может работать.

Код, который никогда не может запустить, называется ** Dead Code **.
В целом, мертвый код не причиняет никакого вреда, но часто указывает на недоразумение, и это может сбивать с толку, что кто -то пытается понять программу.

## постепенное развитие

Когда вы пишете большие функции, вы можете потратить больше
время отладки.
Чтобы справиться со все более сложными программами, вы, возможно, захотите попробовать ** инкрементную разработку **, что является способом добавления и тестирования лишь небольшого количества кода за раз.

В качестве примера, предположим, вы хотите найти расстояние между двумя точками, представленными координатами $ (x_1, Y_1) $ и $ (x_2, y_2) $.
По питагорской теореме расстояние:

$$ \ mathrm {distance} = \ sqrt {(x_2 - x_1)^2 + (y_2 - y_1)^2} $$

Первый шаг - рассмотреть, как должна выглядеть функция `distance`] в Python, то есть каковы входные данные (параметры) и каков выход (возвращаемое значение)?

Для этой функции входные данные являются координатами точек.
Возвратное значение - это расстояние.
Сразу же вы можете написать схему функции:

```python
def distance(x1, y1, x2, y2):
    return 0.0
```

Эта версия пока не вычисляет расстояния - она всегда возвращает ноль.
Но это полная функция с возвратом значения, что означает, что вы можете проверить ее, прежде чем сделать ее более сложной.

Чтобы проверить новую функцию, мы будем называть ее примером аргументов:

```python
distance(1, 2, 4, 6)
```

Я выбрал эти значения так, чтобы горизонтальное расстояние было `3` и
Вертикальное расстояние `4`.
Таким образом, результатом является `5`, гипотенуза `3-4-5` правого треугольника.При тестировании функции полезно знать правильный ответ.

На этом этапе мы подтвердили, что функция работает и возвращает значение, и мы можем начать добавлять код в тело.
Хороший следующий шаг - найти различия `x2 - x1` и `y2 - y1`.
Вот версия, которая хранит эти значения во временных переменных и отображает их.

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    print('dx is', dx)
    print('dy is', dy)
    return 0.0
```

Если функция работает, она должна отображать `dx is 3` и `dy is 4`.
Если так, мы знаем, что функция получает правильные аргументы и
Выполнение первого вычисления правильно.Если нет, то есть только несколько
строки для проверки.

```python
distance(1, 2, 4, 6)
```

До сих пор хорошо.Затем мы рассчитываем сумму квадратов `dx` и `dy`:

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    dsquared = dx**2 + dy**2
    print('dsquared is: ', dsquared)
    return 0.0
```

Опять же, мы можем запустить функцию и проверить выход, который должен быть `25`.

```python
distance(1, 2, 4, 6)
```

Наконец, мы можем использовать `math.sqrt` для вычисления расстояния:

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    dsquared = dx**2 + dy**2
    result = math.sqrt(dsquared)
    print("result is", result)
```

И проверить это.

```python
distance(1, 2, 4, 6)
```

Результат верен, но эта версия функции отображает результат, а не возвращает его, поэтому возвращаемое значение `None`.

Мы можем исправить это, заменив функцию `print`] оператором `return`].

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    dsquared = dx**2 + dy**2
    result = math.sqrt(dsquared)
    return result
```

Эта версия `distance` является чистой функцией.
Если мы называем это так, отображается только результат.

```python
distance(1, 2, 4, 6)
```

И если мы назначаем результат переменной, ничего не отображается.

```python
d = distance(1, 2, 4, 6)
```

Утверждения `print`, которые мы написали, полезны для отладки, но как только функция работает, мы можем их удалить.
Подобный код называется ** леса **, потому что он полезен для создания программы, но не является частью конечного продукта.

Этот пример демонстрирует постепенное развитие.
Ключевыми аспектами этого процесса являются:

1. Начните с рабочей программы, внесите небольшие изменения и проверяйте после каждого изменения.

2. Используйте переменные для сохранения промежуточных значений, чтобы вы могли отобразить и проверить их.

3. Как только программа работает, удалите леса.

В любой момент, если есть ошибка, у вас должна быть хорошая идея, где она находится.
Польшемная разработка может сэкономить вам много времени отладки.

## логические функции

Функции могут возвращать логические значения `True` и `False`, что часто удобно для инкапсулирования сложного теста в функцию.
Например, `is_divisible` проверяет, делится ли `x` на `y` без остатков.

```python
def is_divisible(x, y):
    if x % y == 0:
        return True
    else:
        return False
```

Вот как мы его используем.

```python
is_divisible(6, 4)
```

```python
is_divisible(6, 3)
```

Внутри функции, результатом оператора `==`] является логический, поэтому мы можем написать
функционируйте более кратко, возвращая его напрямую.

```python
def is_divisible(x, y):
    return x % y == 0
```

Логические функции часто используются в условных утверждениях.

```python
if is_divisible(6, 2):
    print('divisible')
```

Может быть заманчиво написать что -то вроде этого:

```python
if is_divisible(6, 2) == True:
    print('divisible')
```

Но сравнение не нужно.

## Recursion с возвращающими значениями

Теперь, когда мы можем писать функции с возвращающимися значениями, мы можем писать рекурсивные функции с возвращающими значениями, и с этой способностью мы прошли важный порог - подмножество Python, которое у нас есть, является ** Turing **, что означает, что мы можем выполнить любые вычисления, которое может быть описано алгоритмом.

Чтобы продемонстрировать рекурсию с возвратными значениями, мы оценим несколько рекурсивно определенных математических функций.
Рекурсивное определение похоже на круговое определение, в том смысле, что определение относится к определению.По -настоящему круговое определение не очень полезно:

> ВОРПОЛ: прилагательное, используемое для описания чего -то, что является Ворпольным.

Если вы видели это определение в словаре, вы можете быть раздражены.
С другой стороны, если вы посмотрите на определение факториальной функции, обозначенное символом $! $, Вы можете получить что -то вроде этого:

$$ \ begin {выровнен}
0!& = 1 \\
n!& = n ~ (n-1)!
\ end {выровнен} $$

В этом определении говорится, что фактор в размере $ 0 $ составляет 1 доллар США, а фактор любой другой стоимости, $ n $,-это $ n $, умноженная на фактор $ N-1 $.

Если вы можете написать рекурсивное определение чего -либо, вы можете написать программу Python, чтобы оценить ее.
После постепенного процесса разработки мы начнем с функции, которая принимает `n` в качестве параметра и всегда возвращается `0`.

```python
def factorial(n):
    return 0
```

Теперь давайте добавим первую часть определения - если аргумент оказался `0`, все, что нам нужно сделать, это вернуть `1`:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return 0
```

Теперь давайте заполним вторую часть - если `n` не является `0`, мы должны сделать рекурсивный
Призовите, чтобы найти фактор `n-1`, а затем умножить результат на `n`:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        recurse = factorial(n-1)
        return n * recurse
```

Поток выполнения для этой программы аналогичен потоку `countdown` в главе 5.
Если мы позвоним `factorial` со значением `3`:

Поскольку `3` не является `0`, мы принимаем вторую ветвь и рассчитываем фактор
из `n-1` \ ...

> Поскольку `2` не является `0`, мы принимаем вторую ветвь и рассчитываем
> Фактор `n-1` \ ...
>
>> Поскольку `1` не является `0`, мы принимаем вторую ветвь и рассчитываем
>> Фактор `n-1` \ ...
>>
>>> Поскольку `0`] равняется `0`, мы принимаем первую ветвь и возвращаем `1` без
>>> Создание еще рекурсивных вызовов.
>>
>> возвращаемое значение, `1`], умножается на `n`, что является `1` и
>> Результат возвращается.
>
> Возвращаемое значение, `1`, умножается на `n`, что `2`, и результат
> возвращается.

Возвращаемое значение `2`] умножается на `n`, что `3`, и результат,
`6`, становится возвращаемым значением вызова функции, который начал целое
процесс.

На следующем рисунке показана диаграмма стека для этой последовательности функций вызовов.

```python
from diagram import Frame, Stack, make_binding

main = Frame([], name='__main__', loc='left')
frames = [main]

ns = 3, 2, 1
recurses = 2, 1, 1
results = 6, 2, 1

for n, recurse, result in zip(ns, recurses, results):
    binding1 = make_binding('n', n)
    binding2 = make_binding('recurse', recurse)
    frame = Frame([binding1, binding2], 
                  name='factorial', value=result,
                  loc='left', dx=1.2)
    frames.append(frame)
    
binding1 = make_binding('n', 0)
frame = Frame([binding1], name='factorial', value=1, 
              shim=1.2, loc='left', dx=1.4)
frames.append(frame)

stack = Stack(frames, dy=-0.45)
```

```python
from diagram import diagram, adjust

width, height, x, y = [2.74, 2.26, 0.73, 2.05]
ax = diagram(width, height)
bbox = stack.draw(ax, x, y)
# adjust(x, y, bbox)
```

Показаны возвращаемые значения, передаваемые обратно в стек.
В каждом кадре возвратное значение является продуктом `n` и `recurse`.

В последнем кадре локальная переменная `recurse` не существует, потому что ветвь, которая создает ее, не работает.

## прыжок веры

Следуя потоку исполнения является одним из способов чтения программ, но он может быстро стать ошеломляющим.Альтернатива - это то, что я называю «прыжком веры».Когда вы приходите к вызову функции, вместо того, чтобы следить за потоком выполнения, вы * предполагаете *, что функция работает правильно и возвращает правильный результат.

На самом деле, вы уже практикуете этот прыжок веры, когда используете встроенные функции.
Когда вы звоните `abs` или `math.sqrt`, вы не изучаете тела этих функций - вы просто предполагаете, что они работают.

То же самое верно, когда вы называете одну из своих собственных функций.Например, ранее мы написали функцию, называемую `is_divisible`, которая определяет, делится ли одно число на другое.Как только мы убедимся в том, что эта функция верна, мы можем использовать ее, не глядя на тело снова.

То же самое относится и к рекурсивным программам.
Когда вы добираетесь до рекурсивного вызова, вместо того, чтобы следить за потоком исполнения, вы должны предположить, что рекурсивный вызов работает, а затем спросить себя: «Предполагая, что я могу вычислить фактор $ n-1 $, могу ли я вычислить фактор $ n $?»
Рекурсивное определение факториального подразумевает, что вы можете, умножив на $ n $.

Конечно, немного странно предположить, что функция работает правильно, когда вы не закончили писать ее, но именно поэтому она называется прыжком веры!

## Фибонааччи

После `factorial` наиболее распространенным примером рекурсивной функции является `fibonacci`, который имеет следующее определение:

$$ \ begin {выровнен}
\ mathrm {fibonaci} (0) & = 0 \\
\ mathrm {fibonaci} (1) & = 1 \\
\ mathrm {fibonaci} (n) & = \ mathrm {fibonacci} (n-1) + \ mathrm {fibonaci} (n-2)
\ ennd {выровнен} $$

Переведено на Python, выглядит так:

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif  n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

Если вы попытаетесь следовать потоку выполнения здесь, даже для небольших значений $ n $, ваша голова взрывается.
Но в соответствии с прыжком веры, если вы предполагаете, что два рекурсивных звонка работают правильно, вы можете быть уверены, что последнее утверждение `return` является правильным.

Кроме того, этот способ вычисления чисел Фибоначчи очень неэффективен.
В [Chapter 10](section_memos) я объясню, почему и предложу способ улучшить его.

## Проверка типов

Что произойдет, если мы позвоним `factorial` и дадим его `1.5` в качестве аргумента?

```python
%%expect RecursionError

factorial(1.5)
```

Это похоже на бесконечную рекурсию.Как это может быть?Функция имеет базовые случаи, когда `n == 1` или `n == 0`.
Но если `n` не является целым числом, мы можем * пропустить * базовый случай и повторяться навсегда.

В этом примере начальное значение `n` составляет `1.5`.
В первом рекурсивном вызове значение `n` составляет `0.5`.
В следующем, это `-0.5`.
Оттуда он становится меньше (более отрицательно), но никогда не будет `0`.

Чтобы избежать бесконечной рекурсии, мы можем использовать встроенную функцию `isinstance`, чтобы проверить тип аргумента.
Вот как мы проверяем, является ли значение целым числом.

```python
isinstance(3, int)
```

```python
isinstance(1.5, int)
```

Теперь вот версия `factorial` с проверкой ошибок.

```python
def factorial(n):
    if not isinstance(n, int):
        print('factorial is only defined for integers.')
        return None
    elif n < 0:
        print('factorial is not defined for negative numbers.')
        return None
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

Сначала он проверяет, является ли `n` целым числом.
Если нет, он отображает сообщение об ошибке и возвращает `None`.

```python
factorial('crunchy frog')
```

Затем он проверяет, является ли отрицательным.
Если так, он отображает сообщение об ошибке и возвращает `None.`

```python
factorial(-2)
```

Если мы пройдем оба чека, мы знаем, что `n` является неотрицательным целым числом, поэтому мы можем быть уверены, что рекурсия прекратится.
Проверка параметров функции, чтобы убедиться, что они имеют правильные типы и значения, называются ** Входная проверка **.

## отладка

Разрыв большой программы на более мелкие функции создает естественные контрольные точки для отладки.
Если функция не работает, есть три возможности для рассмотрения:

- Что -то не так с аргументами, которые выполняет функция, то есть предварительное условие нарушается.

- Что -то не так с этой функцией, то есть пост -кондиционирование нарушается.

- Абонент делает что -то не так с возвратной стоимостью.

Чтобы исключить первую возможность, вы можете добавить оператор `print` в начале функции, которая отображает значения параметров (и, возможно, их типы).
Или вы можете написать код, который явно проверяет предпосылки.

Если параметры выглядят хорошо, вы можете добавить оператор `print` перед каждым оператором `return` и отобразить возвращаемое значение.
Если возможно, вызовите функцию с аргументами, которые позволяют легко проверить результат.

Если функция, кажется, работает, посмотрите на вызов функции, чтобы убедиться, что возвращаемое значение используется правильно - или вообще используется!

Добавление операторов `print` в начале и конце функции может помочь сделать поток выполнения более заметным.
Например, вот версия `factorial` с печатными операторами:

```python
def factorial(n):
    space = ' ' * (4 * n)
    print(space, 'factorial', n)
    if n == 0:
        print(space, 'returning 1')
        return 1
    else:
        recurse = factorial(n-1)
        result = n * recurse
        print(space, 'returning', result)
        return result
```

`space` - это строка космических символов, которая управляет отступом
вывод.Вот результат `factorial(3)`:

```python
factorial(3)
```

Если вы смущены потоком выполнения, этот вид выводов может быть полезным.
Требуется некоторое время, чтобы разработать эффективные леса, но немного лесов может сэкономить много отладки.

## Глоссарий

** возвращаемое значение: **
Результат функции.Если вызов функции используется в качестве выражения, возвращаемое значение - это значение выражения.

** Чистая функция: **
Функция, которая ничего не отображает или не имеет какого -либо другого эффекта, кроме возврата возвращаемого значения.


** Мертвый код: **
Часть программы, которая никогда не может работать, часто потому, что она появляется после оператора `return`.

** Дополнительное развитие: **
План разработки программы, предназначенный для того, чтобы избежать отладки путем добавления и тестирования лишь небольшого количества кода за раз.

** Строительные леса: **
Код, который используется во время разработки программы, но не является частью окончательной версии.

** Turing Complete: **
Язык, или подмножество языка, является полным, если он может выполнить любые вычисления, которые могут быть описаны алгоритмом.

** Входная проверка: **
Проверка параметров функции, чтобы убедиться, что они имеют правильные типы и значения
