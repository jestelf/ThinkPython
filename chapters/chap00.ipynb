# Предисловие



## Для кого эта книга?

Если вы хотите научиться программировать, вы пришли по адресу.
Python — один из лучших языков программирования для новичков и одновременно одна из самых востребованных компетенций.

Вы также выбрали удачное время, потому что сейчас учиться программировать, вероятно, легче, чем когда-либо.
С виртуальными ассистентами вроде ChatGPT вам не придётся учиться в одиночку.
На протяжении всей книги я буду предлагать способы, как использовать эти инструменты, чтобы ускорить обучение.

Книга рассчитана прежде всего на тех, кто никогда не программировал, а также на читателей с опытом в других языках программирования.
Если у вас уже есть значительный опыт в Python, первые главы могут показаться слишком медленными.

Одна из сложностей изучения программирования состоит в том, что приходится осваивать *два* языка: сам язык программирования и словарь терминов, которыми мы пользуемся, говоря о программах.
Если изучать только язык программирования, могут возникать трудности с пониманием сообщений об ошибках, чтением документации, общением с другими людьми или использованием виртуальных ассистентов.
Если вы уже немного программировали, но не освоили этот второй язык, надеюсь, эта книга будет вам полезна.


## Цели книги

Пишa эту книгу, я старался аккуратно обращаться с терминологией.
Я определяю каждый термин в момент первого появления.
В конце каждой главы есть глоссарий, где повторяются введённые термины.

Я также стремился быть лаконичным.
Чем меньше умственных усилий требуется на чтение, тем больше их остаётся на программирование.

Но программированию нельзя научиться, лишь читая книгу — нужно практиковаться.
Поэтому в конце каждой главы есть упражнения, в которых вы можете закрепить материал.

Если читать внимательно и регулярно выполнять упражнения, вы будете продвигаться вперёд.
Но предупреждаю: учиться программировать нелегко, и даже опытные программисты порой испытывают разочарование.
По ходу дела я буду предлагать стратегии, которые помогут писать правильные программы и исправлять ошибки.


## Как работать с книгой

Каждая глава строится на предыдущих, поэтому читайте их по порядку и выполняйте упражнения, прежде чем переходить дальше.

В первых шести главах рассматриваются базовые элементы: арифметика, условные операторы и циклы.
Там же вводится важнейшее понятие программирования — функции, а также мощный способ их применения — рекурсия.

В главах 7 и 8 разбираются строки — они могут представлять буквы, слова и предложения — и алгоритмы работы с ними.

В главах с 9 по 12 мы знакомимся с основными структурами данных Python — списками, словарями и кортежами, которые позволяют писать эффективные программы.
В 12 главе приводятся алгоритмы анализа текста и случайного его порождения.
Такие алгоритмы лежат в основе больших языковых моделей (LLM), поэтому эта глава поможет понять, как работают инструменты вроде ChatGPT.

Глава 13 посвящена способам долговременного хранения данных — файлам и базам данных.
В качестве упражнения вы напишете программу, которая ищет в файловой системе дубликаты файлов.

В главах с 14 по 17 вводится объектно-ориентированное программирование (ООП) — способ организации программ и данных, с которыми они работают.
Многие библиотеки Python написаны в объектно-ориентированном стиле, поэтому эти главы помогут понять их устройство и создавать собственные объекты.

Цель этой книги — не охватить весь язык Python.
Я сосредоточился на подмножестве языка, которое даёт наибольшие возможности при минимуме понятий.
Тем не менее у Python есть множество средств для эффективного решения типовых задач.
О некоторых из них рассказывается в 18 главе.

Наконец, в 19 главе я делюсь завершающими мыслями и советами по продолжению изучения программирования.


## Что нового в третьем издании?

Наибольшие изменения в этой версии связаны с двумя новыми технологиями — Jupyter-ноутбуками и виртуальными ассистентами.

Каждая глава этой книги — это Jupyter-ноутбук, документ, содержащий обычный текст и код.
Мне так легче писать код, тестировать его и поддерживать соответствие тексту.
Вам это позволит запускать код, изменять его и выполнять задания в одном месте.
Инструкции по работе с ноутбуками приведены в первой главе.

Другая крупная новинка — советы по использованию виртуальных ассистентов вроде ChatGPT, которые помогут вам учиться быстрее.
Когда предыдущее издание вышло в 2016 году, предшественники этих инструментов были гораздо менее полезны, и большинство людей о них не знали.
Теперь они стали стандартным инструментом разработки программного обеспечения, и, на мой взгляд, станут революционным средством обучения программированию и многому другому.

Остальные изменения книги связаны с моими сожалениями о втором издании.

Во-первых, я недостаточно уделил внимания тестированию программ.
Уже в 2016 году это было упущением, а с появлением виртуальных ассистентов автоматизированное тестирование стало ещё важнее.
Поэтому в этом издании представлены наиболее популярные средства тестирования Python — `doctest` и `unittest`, а также несколько упражнений для практики работы с ними.

Во-вторых, задания во втором издании были неравномерными: одни получились интереснее, другие оказались слишком сложными.
Перевод книги в формат Jupyter-ноутбуков помог мне разработать и проверить более последовательный и эффективный набор упражнений.

В этой редакции последовательность тем почти не изменилась, но я переставил несколько глав и объединил две короткие главы в одну.
Кроме того, я расширил раздел о строках, добавив регулярные выражения.

В нескольких главах используются черепашья графика.
В прежних изданиях я применял модуль `turtle` из стандартной библиотеки, но, к сожалению, он не работает в Jupyter-ноутбуках.
Поэтому я заменил его новым модулем turtle, который должен быть проще в использовании.

Наконец, я переписал значительную часть текста, уточняя и сокращая его там, где это было нужно.

Я очень горжусь этим изданием и надеюсь, оно вам понравится!


## Начало работы

Для большинства языков программирования, включая Python, существует много инструментов для написания и запуска программ.
Эти инструменты называются интегрированными средами разработки (IDE).
В общих чертах IDE бывают двух типов:

* Одни работают с файлами кода и предоставляют средства для их редактирования и запуска.

* Другие в основном работают с ноутбуками — документами, содержащими текст и код.

Новичкам я рекомендую начать с ноутбучной среды разработки, такой как Jupyter.

Ноутбуки этой книги доступны в онлайн‑репозитории по адресу <https://allendowney.github.io/ThinkPython>.

Существует два способа их использовать:

* Можно скачать ноутбуки и запускать их на своём компьютере. В этом случае придётся установить Python и Jupyter; это несложно, но если вы хотите учить Python, бывает досадно тратить много времени на установку ПО.

* Альтернативный вариант — работать с ноутбуками на Colab, это среда Jupyter, которая запускается в веб‑браузере, так что ничего устанавливать не нужно. Colab работает под управлением Google и бесплатен.

Если вы только начинаете, настоятельно рекомендую стартовать с Colab.
