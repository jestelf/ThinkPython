# Классы и методы

Python — это **объектно-ориентированный язык**, то есть он предоставляет возможности, поддерживающие объектно-ориентированное программирование, которое характеризуется следующим:

-   Большая часть вычислений выражается через операции над объектами.

-   Объекты часто представляют вещи из реального мира, а методы обычно соответствуют тому, как эти вещи взаимодействуют.

-   В программах присутствуют определения классов и методов.

Например, в предыдущей главе мы определили класс `Time`, соответствующий тому, как люди записывают время суток, и написали функции, соответствующие действиям, которые люди выполняют со временем.
Но между определением класса `Time` и последующими функциями не было явной связи.
Мы можем сделать эту связь явной, переписав функцию в виде **метода**, определённого внутри класса.

## Определение методов

В предыдущей главе мы определили класс `Time` и написали функцию `print_time`, которая выводит время суток.

```python
class Time:
    """Represents the time of day."""

def print_time(time):
    s = f'{time.hour:02d}:{time.minute:02d}:{time.second:02d}'
    print(s)
```

Чтобы превратить `print_time` в метод, достаточно перенести определение функции внутрь определения класса. Обратите внимание на изменение отступов.

Одновременно мы изменим имя параметра с `time` на `self`. Это изменение не обязательно, но принято называть первым параметром метода `self`.

```python
class Time:
    """Represents the time of day."""    

    def print_time(self):
        s = f'{self.hour:02d}:{self.minute:02d}:{self.second:02d}'
        print(s)
```

Чтобы вызвать этот метод, нужно передать объект `Time` в качестве аргумента.
Вот функция, с помощью которой мы будем создавать объект `Time`.

```python
def make_time(hour, minute, second):
    time = Time()
    time.hour = hour
    time.minute = minute
    time.second = second
    return time
```

А вот и экземпляр `Time`.

```python
start = make_time(9, 40, 0)
```

Теперь `print_time` можно вызвать двумя способами. Первый (менее распространённый) — использовать синтаксис функции.

```python
Time.print_time(start)
```

В этой версии `Time` — имя класса, `print_time` — имя метода, а `start` передаётся в качестве параметра.
Второй (более идиоматичный) способ — использовать синтаксис метода:

```python
start.print_time()
```

В этой версии `start` — объект, на котором вызывается метод; его называют **приёмником**, по аналогии с тем, что вызов метода подобен отправке сообщения объекту.

Независимо от синтаксиса поведение метода одинаково. Приёмник присваивается первому параметру, поэтому внутри метода `self` ссылается на тот же объект, что и `start`.

## Ещё один метод

Вот функция `time_to_int` из предыдущей главы.

```python
def time_to_int(time):
    minutes = time.hour * 60 + time.minute
    seconds = minutes * 60 + time.second
    return seconds
```

А вот вариант, переписанный в виде метода.

```python
%%add_method_to Time

    def time_to_int(self):
        minutes = self.hour * 60 + self.minute
        seconds = minutes * 60 + self.second
        return seconds
```

Первая строка использует специальную команду `add_method_to`, которая добавляет метод к уже определённому классу.
Эта команда работает в Jupyter Notebook, но не является частью Python и потому не будет работать в других средах.
Обычно все методы класса находятся внутри определения класса и определяются одновременно с ним.
Но для этой книги полезно определять по одному методу за раз.

Как и в предыдущем примере, определение метода имеет отступ, а имя параметра — `self`.
В остальном метод идентичен функции.
Вот как его вызвать.

```python
start.time_to_int()
```

Часто говорят, что функцию мы "вызываем", а метод "инициируем", но по сути это одно и то же.

## Статические методы

Рассмотрим ещё функцию `int_to_time`. Вот её вариант из предыдущей главы.

```python
def int_to_time(seconds):
    minute, second = divmod(seconds, 60)
    hour, minute = divmod(minute, 60)
    return make_time(hour, minute, second)
```

Эта функция принимает параметр `seconds` и возвращает новый объект `Time`.
Если преобразовать её в метод класса `Time`, его придётся вызывать на объекте `Time`.
Но если мы хотим создать новый объект `Time`, на чём мы должны его вызывать?

Эту проблему курицы и яйца можно решить с помощью **статического метода**, который не требует экземпляра класса для вызова.
Вот как переписать эту функцию как статический метод.

```python
%%add_method_to Time

    def int_to_time(seconds):
        minute, second = divmod(seconds, 60)
        hour, minute = divmod(minute, 60)
        return make_time(hour, minute, second)
```

Поскольку это статический метод, у него нет параметра `self`.
Вызываем его с помощью `Time`, то есть объекта класса.

```python
start = Time.int_to_time(34800)
```

Результатом будет новый объект, представляющий 9:40.

```python
start.print_time()
```

Теперь, имея `Time.from_seconds`, мы можем написать `add_time` как метод.
Вот функция из предыдущей главы.

```python
def add_time(time, hours, minutes, seconds):
    duration = make_time(hours, minutes, seconds)
    seconds = time_to_int(time) + time_to_int(duration)
    return int_to_time(seconds)
```

А вот версия, переписанная в виде метода.

```python
%%add_method_to Time

    def add_time(self, hours, minutes, seconds):
        duration = make_time(hours, minutes, seconds)
        seconds = time_to_int(self) + time_to_int(duration)
        return Time.int_to_time(seconds)
```

У метода `add_time` есть параметр `self`, потому что это не статический, а обычный **экземплярный метод**.
Чтобы вызвать его, нам нужен экземпляр `Time`.

```python
end = start.add_time(1, 32, 0)
print_time(end)
```

## Сравнение объектов Time

В качестве ещё одного примера напишем `is_after` как метод.
Вот функция `is_after`, которая была решением упражнения в предыдущей главе.

```python
def is_after(t1, t2):
    return time_to_int(t1) > time_to_int(t2)
```

А вот она же в виде метода.

```python
%%add_method_to Time

    def is_after(self, other):
        return self.time_to_int() > other.time_to_int()
```

Поскольку мы сравниваем два объекта, а первым параметром является `self`, второй параметр назовём `other`.
Чтобы использовать этот метод, нужно вызвать его на одном объекте и передать
другой в качестве аргумента.

```python
end.is_after(start)
```

Хорошо ещё то, что этот синтаксис почти читается как вопрос: "`end` позже `start`?"

## Метод `__str__`

Создавая метод, можно выбрать почти любое имя. Однако некоторые имена имеют особое значение.
Например, если у объекта есть метод `__str__`, Python использует его для преобразования объекта в строку.
Вот пример метода `__str__` для объекта времени.

```python
%%add_method_to Time

    def __str__(self):
        s = f'{self.hour:02d}:{self.minute:02d}:{self.second:02d}'
        return s
```

Этот метод похож на `print_time` из предыдущей главы, но возвращает строку вместо вывода её на экран.

Вызвать его можно обычным способом.

```python
end.__str__()
```

Но Python может вызвать его и сам.
Если использовать встроенную функцию `str` для преобразования объекта `Time` в строку, Python применит метод `__str__` класса `Time`.

```python
str(end)
```

То же самое произойдёт, если напечатать объект `Time`.

```python
print(end)
```

Методы вроде `__str__` называются **специальными**. Узнать их можно по тому, что они начинаются и заканчиваются двумя символами подчёркивания.

## Метод `__init__`

Самым особенным из специальных методов является `__init__`, потому что он инициализирует атрибуты нового объекта.
Метод `__init__` для класса `Time` может выглядеть так:

```python
%%add_method_to Time

    def __init__(self, hour=0, minute=0, second=0):
        self.hour = hour
        self.minute = minute
        self.second = second
```

Теперь при создании объекта `Time` Python вызывает `__init__` и передаёт ему аргументы.
Так мы можем создать объект и сразу инициализировать его атрибуты.

```python
time = Time(9, 40, 0)
print(time)
```

В этом примере параметры необязательны, поэтому если вызвать `Time` без аргументов,
получатся значения по умолчанию.

```python
time = Time()
print(time)
```

Если передать один аргумент, он заменит значение `hour`:

```python
time = Time(9)
print(time)
```

Если передать два аргумента, они заменят `hour` и `minute`.

```python
time = Time(9, 45)
print(time)
```

А если передать три аргумента, они заменят все три значения по умолчанию.

Когда я пишу новый класс, почти всегда начинаю с `__init__`, который упрощает создание объектов, и `__str__`, который помогает при отладке.

## Перегрузка операторов

Определив другие специальные методы, можно задать поведение операторов для пользовательских типов. Например, если определить метод `__add__` для класса `Time`, можно использовать оператор `+` с объектами `Time`.

Вот метод `__add__`.

```python
%%add_method_to Time

    def __add__(self, other):
        seconds = self.time_to_int() + other.time_to_int()
        return Time.int_to_time(seconds)
```

Использовать его можно так.

```python
duration = Time(1, 32)
end = start + duration
print(end)
```

При выполнении этих трёх строк кода происходит много всего:

* При создании объекта `Time` вызывается метод `__init__`.

* При использовании оператора `+` с объектом `Time` вызывается его метод `__add__`.

* А при печати объекта `Time` вызывается метод `__str__`.

Изменение поведения оператора так, чтобы он работал с пользовательскими типами, называется **перегрузкой операторов**.
Для каждого оператора, например `+`, существует соответствующий специальный метод, например `__add__`.

## Отладка

Объект `Time` считается корректным, если значения `minute` и `second` находятся в диапазоне от `0` до `60` — включая `0`, но не `60` — и если `hour` положительно.
Кроме того, `hour` и `minute` должны быть целыми числами, хотя `second` может иметь дробную часть.
Такие требования называются **инвариантами**, потому что они должны быть истинны всегда. Если это не так, значит что-то пошло не так.

Проверка инвариантов помогает обнаруживать ошибки и находить их причины. Например, можно написать метод `is_valid`, который принимает объект Time и возвращает `False`, если тот нарушает инвариант.

```python
%%add_method_to Time

    def is_valid(self):
        if self.hour < 0 or self.minute < 0 or self.second < 0:
            return False
        if self.minute >= 60 or self.second >= 60:
            return False
        if not isinstance(self.hour, int):
            return False
        if not isinstance(self.minute, int):
            return False
        return True
```

Затем в начале каждого метода можно проверять аргументы на корректность.

```python
%%add_method_to Time

    def is_after(self, other):
        assert self.is_valid(), 'self is not a valid Time'
        assert other.is_valid(), 'self is not a valid Time'
        return self.time_to_int() > other.time_to_int()
```

Оператор `assert` вычисляет следующее за ним выражение. Если результат `True`, ничего не происходит; если `False`, возбуждается `AssertionError`.
Вот пример.

```python
duration = Time(minute=132)
print(duration)
```

```python
%%expect AssertionError

start.is_after(duration)
```

Выражения `assert` полезны тем, что отделяют код, работающий с нормальными условиями, от кода, проверяющего наличие ошибок.

## Словарь терминов

**object-oriented language:**
Язык, предоставляющий средства для объектно-ориентированного программирования, прежде всего пользовательские типы.

**method:**
Функция, определённая внутри класса и вызываемая для экземпляров этого класса.

**receiver:**
Объект, на котором вызывается метод.

**static method:**
Метод, который можно вызвать без объекта-приёмника.

**instance method:**
Метод, который должен вызываться с объектом-приёмником.

**special method:**
Метод, изменяющий работу операторов и некоторых функций с объектом.

**operator overloading:**
Процесс изменения работы операторов с пользовательскими типами с помощью специальных методов.

**invariant:**
Условие, которое должно оставаться истинным во время выполнения программы.
