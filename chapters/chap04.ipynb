## Модуль jupyturtle

Чтобы использовать модуль `jupyturtle`, импортируем его так.

```python
import jupyturtle
```

Теперь можно использовать функции из этого модуля, например `make_turtle` и `forward`.

```python
jupyturtle.make_turtle()
jupyturtle.forward(100)
```

`make_turtle` создаёт **холст** — область на экране, где можно рисовать, и черепаху, которая изображается кругом‑панцирем и треугольной головой. Круг показывает положение черепахи, а треугольник указывает направление, куда она «смотрит».

`forward` перемещает черепаху на указанное расстояние по направлению её движения, оставляя за собой отрезок линии. Расстояние измеряется в произвольных единицах — фактический размер зависит от вашего экрана.

Функции из модуля `jupyturtle` мы будем использовать много раз, поэтому будет удобнее не писать имя модуля каждый раз. Это возможно, если импортировать модуль так.

```python
from jupyturtle import make_turtle, forward
```

Такая версия инструкции импортирует `make_turtle` и `forward` из модуля `jupyturtle`, так что их можно вызывать вот так.

```python
make_turtle()
forward(100)
```

Модуль `jupyturtle` предоставляет ещё две функции, которые нам пригодятся, — `left` и `right`. Импортируем их так.

```python
from jupyturtle import left, right
```

`left` заставляет черепаху повернуть налево. Функция принимает один аргумент — угол поворота в градусах. Например, так можно выполнить поворот на 90 градусов.

```python
make_turtle()
forward(50)
left(90)
forward(50)
```

Эта программа перемещает черепаху сначала на восток, затем на север, оставляя позади две линии. Прежде чем продолжить, попробуйте изменить предыдущую программу так, чтобы она рисовала квадрат.

## Рисуем квадрат

Вот один из способов нарисовать квадрат.

```python
make_turtle()

forward(50)
left(90)

forward(50)
left(90)

forward(50)
left(90)

forward(50)
left(90)
```

Поскольку эта программа повторяет одну и ту же пару строк четыре раза, то же самое можно сделать короче с помощью цикла `for`.

```python
make_turtle()
for i in range(4):
    forward(50)
    left(90)
```

## Инкапсуляция и обобщение

Возьмём код из предыдущей части, который рисует квадрат, и поместим его в функцию `square`.

```python
def square():
    for i in range(4):
        forward(50)
        left(90)
```

Теперь её можно вызвать так.

```python
make_turtle()
square()
```

Упаковка участка кода в функцию называется **инкапсуляцией**. Одно из её преимуществ в том, что код получает имя, служащее своего рода документацией. Ещё один плюс: если нужно использовать код повторно, проще дважды вызвать функцию, чем копировать её содержимое!

В текущей версии размер квадрата всегда равен `50`. Если мы хотим рисовать квадраты разных размеров, можно передавать длину стороны в качестве параметра.

```python
def square(length):
    for i in range(4):
        forward(length)
        left(90)
```

Теперь мы можем рисовать квадраты разного размера.

```python
make_turtle()
square(30)
square(60)
```

Добавление параметра к функции называется **обобщением**, потому что делает её более универсальной: в предыдущей версии квадрат был всегда одного размера, а в этой — любого.

Если добавить ещё один параметр, функцию можно сделать ещё более общей. Следующая функция рисует правильные многоугольники с заданным числом сторон.

```python
def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)
```

В правильном многоугольнике с `n` сторонами угол между соседними сторонами равен `360 / n` градусов.

Следующий пример рисует 7‑угольник со стороной `30`.

```python
make_turtle()
polygon(7, 30)
```

Когда у функции несколько числовых аргументов, легко забыть, что они означают и в каком порядке их передавать. Поэтому полезно включать имена параметров в список аргументов.

```python
make_turtle()
polygon(n=7, length=30)
```

Такой способ иногда называют «именованными аргументами», потому что в них присутствуют имена параметров. Но в Python их чаще называют **аргументами-ключевыми словами** (не путать с ключевыми словами языка `for` и `def`).

Использование оператора присваивания `=` напоминает, как работают аргументы и параметры: при вызове функции аргументы присваиваются параметрам.

## Аппроксимация окружности

Предположим, мы хотим нарисовать окружность. Это можно сделать приближённо, нарисовав многоугольник с большим числом сторон, так что каждая из них будет достаточно маленькой, чтобы её было трудно увидеть. Вот функция, которая с помощью `polygon` рисует 30‑угольник, приближающий окружность.

```python
import math

def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    polygon(n, length)
```

`circle` принимает радиус окружности как параметр. Она вычисляет `circumference` — длину окружности с заданным радиусом. `n` — число сторон, поэтому `circumference / n` — длина каждой стороны.

Выполнение этой функции может занять много времени. Мы можем ускорить её, вызвав `make_turtle` с именованным аргументом `delay`, который задаёт паузу в секундах после каждого шага. Значение по умолчанию — 0.2 секунды; если установить 0.02, программа будет работать примерно в десять раз быстрее.

```python
make_turtle(delay=0.02)
circle(30)
```

Недостаток этого решения в том, что `n` — константа: для очень больших окружностей стороны получаются слишком длинными, а для маленьких мы зря тратим время, рисуя очень короткие отрезки. Один из вариантов — обобщить функцию, добавив параметр `n`. Но пока оставим всё как есть.

## Рефакторинг

Теперь напишем более общую версию `circle` под названием `arc`, которая принимает второй параметр `angle` и рисует дугу окружности на указанный угол. Например, если `angle` равен 360 градусов, получаем полную окружность; если 180 — полукруг.

При написании `circle` мы смогли использовать `polygon`, потому что многоугольник с большим числом сторон хорошо приближает окружность. Но для `arc` `polygon` не подойдёт.

Вместо этого мы создадим более общую версию `polygon`, которую назовём `polyline`.

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)
```

`polyline` принимает в качестве параметров количество отрезков `n`, длину каждого отрезка `length` и угол между ними `angle`.

Теперь мы можем переписать `polygon`, используя `polyline`.

```python
def polygon(n, length):
    angle = 360.0 / n
    polyline(n, length, angle)
```

И с помощью `polyline` можно написать `arc`.

```python
def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)
```

`arc` похожа на `circle`, но вычисляет `arc_length`, то есть часть длины окружности. Наконец, мы можем переписать `circle`, чтобы она использовала `arc`.

```python
def circle(radius):
    arc(radius,  360)
```

Чтобы убедиться, что функции работают как нужно, используем их для рисования чего-то вроде улитки. При `delay=0` черепаха движется максимально быстро.

```python
make_turtle(delay=0)
polygon(n=20, length=9)
arc(radius=70, angle=70)
circle(radius=10)
```

В этом примере мы начали с рабочего кода и перестроили его, разбив на другие функции. Такие изменения, которые улучшают код, не меняя его поведение, называются **рефакторингом**.

Если бы мы планировали заранее, возможно, сначала написали бы `polyline` и избежали рефакторинга, но часто в начале проекта ещё не ясно, как лучше разделить программу на функции. Начав кодирование, вы лучше понимаете задачу. Иногда рефакторинг показывает, что вы чему‑то научились.

## План разработки

**План разработки** — это процесс написания программ. В этой главе мы использовали метод «инкапсуляция и обобщение». Его шаги таковы:

1. Начните с небольшой программы без определений функций.

2. Как только программа заработает, выделите в ней связную часть, инкапсулируйте её в функции и дайте имя.

3. Обобщите функцию, добавив подходящие параметры.

4. Повторяйте шаги 1–3, пока не получите набор рабочих функций.

5. Ищите возможности улучшить программу рефакторингом. Например, если в нескольких местах есть похожий код, вынесите его в достаточно общую функцию.

У этого подхода есть недостатки — позже мы увидим другие варианты — но он полезен, когда заранее непонятно, как разделить программу на функции. Такой подход позволяет проектировать по ходу дела.

Проектирование функции состоит из двух частей:

* **Интерфейс** определяет, как функция используется: её имя, параметры и что она должна делать.

* **Реализация** описывает, как функция выполняет свою работу.

Например, ниже приведена первая версия `circle`, которая использует `polygon`.

```python
def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    polygon(n, length)
```

А вот переработанная версия, которая использует `arc`.

```python
def circle(radius):
    arc(radius,  360)
```

Эти две функции имеют одинаковый интерфейс — принимают одни и те же параметры и делают одно и то же, — но реализация у них разная.

## Докстроки

**Докстрока** — это строка в начале функции, которая объясняет её интерфейс («doc» сокращение от «documentation»). Вот пример:

```python
def polyline(n, length, angle):
    """Draws line segments with the given length and angle between them.
    
    n: integer number of line segments
    length: length of the line segments
    angle: angle between segments (in degrees)
    """    
    for i in range(n):
        forward(length)
        left(angle)
```

По соглашению, докстроки записываются в тройных кавычках, поэтому их ещё называют **многострочными строками**, так как они могут занимать несколько строк.

Докстрока должна:

* Кратко объяснять, что делает функция, не вдаваясь в детали реализации,

* описывать, как каждый параметр влияет на поведение функции,

* указывать тип каждого параметра, если это не очевидно.

Написание такой документации — важная часть проектирования интерфейса. Хорошо спроектированный интерфейс легко объяснить; если вам тяжело объяснить, что делает функция, возможно, интерфейс стоит улучшить.

## Отладка

Интерфейс похож на контракт между функцией и вызывающей стороной. Вызывающий обещает предоставить определённые аргументы, а функция — выполнить определённую работу.

Например, `polyline` требует три аргумента: `n` должно быть целым числом; `length` — положительным числом; `angle` — числом, подразумевается градусная мера.

Эти требования называют **предусловиями**, потому что они должны выполняться до начала работы функции. Соответственно, условия на момент её окончания — **постусловия**. Постусловия описывают предполагаемый эффект функции (например, рисование отрезков) и возможные побочные эффекты (например, перемещение черепахи или другие изменения).

Предусловия лежат на ответственности вызывающей стороны. Если она их нарушает и функция работает неправильно, ошибка в вызывающем коде, а не в функции.

Если предусловия выполнены, а постусловия — нет, ошибка в функции. Чёткие предусловия и постусловия помогают при отладке.

## Словарь

**interface design:** Процесс проектирования интерфейса функции, включающий определение параметров, которые она должна принимать.

**keyword argument:** Аргумент, в котором указано имя параметра.

**refactoring:** Процесс модификации работающей программы для улучшения интерфейсов функций и других свойств кода.

**precondition:** Требование, которое должно выполняться вызывающей стороной до начала работы функции.

Следующая функция использует `penup` и `pendown`, чтобы переместить черепаху без следа.

```python
from jupyturtle import penup, pendown

def jump(length):
    """Move forward length units without leaving a trail.
    
    Postcondition: Leaves the pen down.
    """
    penup()
    forward(length)
    pendown()
```
