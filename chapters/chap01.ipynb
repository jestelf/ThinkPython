## Арифметические операторы

** Арифметический оператор ** является символом, который представляет арифметическое вычисление.Например, знак плюса, `+`, выполняет дополнение.

```python
30 + 12
```

Знак минус, `-`, является оператором, который выполняет вычитание.

```python
43 - 1
```

Звездочка, `*` выполняет умножение.

```python
6 * 7
```

И прямая смена, `/`, выполняет дивизион:

```python
84 / 2
```

Обратите внимание, что результат деления составляет `42.0`, а не« 42`.Это потому, что в Python есть два типа цифр:

*** Целые числа **, которые представляют цифры без дробной или десятичной части, и

*** Числа с плавающей точкой **, которые представляют целые числа и числа с десятичной точкой.

Если вы добавляете, вычитаете или умножаете два целых числа, результатом является целое число.
Но если вы делите два целых числа, результатом является номер с плавающей запятой.
Python предоставляет еще одного оператора, `//`, который выполняет ** целочисленное разделение **.
Результат целочисленного подразделения всегда является целым числом.

```python
84 // 2
```

Целочисленное подразделение также называется «Полевой дивизией», потому что он всегда окружает (к «полу»).

```python
85 // 2
```

Наконец, оператор `**` выполняет экспонентацию;то есть он поднимает
номер к силе:

```python
7 ** 2
```

На некоторых других языках каретка, `^`, используется для экспоненты, но в Python
Это кусочек оператора под названием XOR.
Если вы не знакомы с битой операторов, результат может быть неожиданным:

```python
7 ^ 2
```

Я не буду охватывать битовые операторы в этой книге, но вы можете прочитать о
их на <http://wiki.python.org/moin/bitiseoperators>.

## выражения

Сборник операторов и цифр называется ** выражением **.
Выражение может содержать любое количество операторов и чисел.
Например, вот выражение, которое содержит два оператора.

```python
6 + 6 ** 2
```

Обратите внимание, что экспонентация происходит до добавления.
Python следует по порядку операций, который вы могли выучить в математическом классе: экспонентация происходит до размножения и деления, которые происходят до добавления и вычитания.

В следующем примере умножение происходит до добавления.

```python
12 + 5 * 6
```

Если вы хотите, чтобы добавление произошло первым, вы можете использовать скобки.

```python
(12 + 5) * 6
```

Каждое выражение имеет ** значение **.
Например, выражение `6 * 7` имеет значение` 42`.

## формальные и естественные языки

** Природные языки ** - это языки, которые люди говорят, как на английском, испанском и французском языках.Они не были спроектированы людьми;Они развивались естественно.

** Формальные языки ** - это языки, которые разработаны людьми для конкретных применений.
Например, нотация, которую используют математики, является формальным языком, который особенно хорош в обозначении отношений между числами и символами.
Точно так же языки программирования - это формальные языки, которые были разработаны для выражения вычислений.

Хотя формальные и природные языки имеют некоторые особенности в
Общие существуют важные различия:

* Неоднозначность: естественные языки полны двусмысленности, с которыми сталкиваются люди
Использование контекстуальных подсказок и другой информации.Формальные языки есть
спроектировано, чтобы быть почти или совершенно однозначным, что означает, что
Любая программа имеет ровное значение, независимо от контекста.

* Избыточность: чтобы компенсировать двусмысленность и уменьшить недоразумения,
Природные языки используют избыточность.В результате они
часто многослов.Формальные языки менее избыточные и более краткие.

* Литеральная буква: естественные языки полны идиомы и метафоры.Формальные языки означают именно то, что они говорят.

Поскольку мы все расти, говоря о естественных языках, иногда трудно приспособиться к формальным языкам.
Формальные языки более плотные, чем естественные языки, поэтому для их чтения требуется больше времени.
Кроме того, структура важна, поэтому не всегда лучше читать сверху вниз, слева направо.
Наконец, детали имеют значение.Небольшие ошибки в орфографии и
пунктуация, с которой вы можете сойти с рук на естественных языках, может сделать
Большая разница в формальном языке.
