# Функции

В предыдущей главе мы использовали несколько функций Python, таких как `int` и `float`, а также некоторые из модуля `math`, например `sqrt` и `pow`.
В этой главе вы узнаете, как создавать свои собственные функции и выполнять их.

Эти простые примеры демонстрируют важную особенность — возможность писать свои функции является основой программирования.

Эта глава также вводит новый оператор — цикл `for`, который используется для повторения вычислений.

## Определение новых функций

**Определение функции** задаёт имя новой функции и последовательность инструкций, которые выполняются при её вызове. Вот пример:

````python
def print_lyrics():
    print("I'm a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")
````

`def` — это ключевое слово, которое указывает на определение функции.
Имя функции — `print_lyrics`.
Любое допустимое имя переменной также является допустимым именем функции.

Пустые скобки после имени показывают, что функция не принимает аргументов.

Первая строка определения функции называется **заголовком**, остальная часть — **телом**.
Заголовок должен заканчиваться двоеточием, а тело должно быть с отступом. По соглашению отступ составляет четыре пробела. 
Тело этой функции состоит из двух инструкций `print`; в целом тело функции может содержать любое количество инструкций любого типа.

Определение функции создает **функциональный объект**, который можно вывести вот так.

````python
print_lyrics
````

Результат показывает, что `print_lyrics` — функция без аргументов.
`__main__` — имя модуля, содержащего `print_lyrics`.

Теперь, когда функция определена, мы можем вызывать её так же, как встроенные функции.

````python
print_lyrics()
````

Когда функция выполняется, она исполняет инструкции в теле, которые выводят первые две строки из «Песни дровосека».

## Параметры

Некоторые из функций, которые мы уже видели, требуют аргументы; например, при вызове `abs` вы передаёте число.
Некоторые функции принимают больше одного аргумента; например, `math.pow` принимает два — основание и показатель степени.

Вот определение функции, которая принимает аргумент.

````python
def print_twice(string):
    print(string)
    print(string)
````

Имя переменной в скобках называется **параметром**.
Когда функция вызывается, значение аргумента присваивается параметру.
Например, мы можем вызвать `print_twice` так.

````python
print_twice('Dennis Moore, ')
````

Выполнение этой функции имеет тот же эффект, что и присвоение аргумента параметру с последующим выполнением тела функции, как здесь.

````python
string = 'Dennis Moore, '
print(string)
print(string)
````

В качестве аргумента можно использовать и переменную.

````python
line = 'Dennis Moore, '
print_twice(line)
````

В этом примере значение `line` присваивается параметру `string`.

## Вызов функций

После того как вы определили функцию, её можно использовать внутри другой функции.
Чтобы показать это, напишем функции, которые выводят текст песни «The Spam Song» (<https://www.songfacts.com/lyrics/monty-python/the-spam-song>).

> Spam, Spam, Spam, Spam,
> Spam, Spam, Spam, Spam,
> Spam, Spam,
> (Lovely Spam, Wonderful Spam!)
> Spam, Spam,

Начнём со следующей функции, которая принимает два параметра.


````python
def repeat(word, n):
    print(word * n)
````

Мы можем использовать эту функцию, чтобы вывести первую строку песни, вот так.

````python
spam = 'Spam, '
repeat(spam, 4)
````

Чтобы вывести первые две строки, мы можем определить новую функцию, использующую `repeat`.

````python
def first_two_lines():
    repeat(spam, 4)
    repeat(spam, 4)
````

А затем вызвать её вот так.

````python
first_two_lines()
````

Чтобы вывести последние три строки, мы можем определить ещё одну функцию, которая также использует `repeat`.

````python
def last_three_lines():
    repeat(spam, 2)
    print('(Lovely Spam, Wonderful Spam!)')
    repeat(spam, 2)
````

````python
last_three_lines()
````

Наконец, мы можем собрать всё вместе в одной функции, которая печатает весь куплет.

````python
def print_verse():
    first_two_lines()
    last_three_lines()
````

````python
print_verse()
````

Когда мы вызываем `print_verse`, эта функция вызывает `first_two_lines`, которая вызывает `repeat`, а та вызывает `print`.
Это много функций.

Конечно, то же самое можно было сделать и меньшим количеством функций, но цель примера — показать, как функции могут работать вместе.

## Повторение

Если мы хотим вывести больше одного куплета, можно использовать оператор `for`.
Вот простой пример.

````python
for i in range(2):
    print(i)
````

Первая строка — это заголовок, который заканчивается двоеточием.
Вторая строка — тело цикла, которое должно быть с отступом.

Заголовок начинается с ключевого слова `for`, новой переменной `i` и другого ключевого слова `in`. 
Он использует функцию `range`, чтобы создать последовательность из двух значений — `0` и `1`.
В Python при счёте мы обычно начинаем с `0`.

Когда выполняется оператор `for`, он присваивает первое значение из `range` переменной `i`, а затем запускает функцию `print` в теле цикла, которая выводит `0`.

Дойдя до конца тела, он возвращается к заголовку, поэтому такой оператор называют **циклом**.
Во второй проход цикла переменной `i` присваивается следующее значение из `range`, и оно выводится.
Затем, так как это последнее значение из `range`, цикл завершается.

Вот как можно использовать цикл `for`, чтобы вывести два куплета песни.

````python
for i in range(2):
    print("Verse", i)
    print_verse()
    print()
````

Вы можете поместить цикл `for` внутрь функции.
Например, `print_n_verses` принимает параметр `n`, который должен быть целым числом, и выводит заданное количество куплетов. 

````python
def print_n_verses(n):
    for i in range(n):
        print_verse()
        print()
````

В этом примере мы не используем `i` в теле цикла, но имя переменной в заголовке всё равно должно быть.

## Переменные и параметры являются локальными

Когда вы создаёте переменную внутри функции, она **локальная**,
то есть существует только внутри этой функции.
Например, следующая функция принимает два аргумента, объединяет их и печатает результат дважды.

````python
def cat_twice(part1, part2):
    cat = part1 + part2
    print_twice(cat)
````

Вот пример его использования:

````python
line1 = 'Always look on the '
line2 = 'bright side of life.'
cat_twice(line1, line2)
````

Когда выполняется `cat_twice`, создаётся локальная переменная `cat`, которая уничтожается после завершения функции.
Если попытаться вывести её, мы получим `NameError`:

````python
%%expect NameError

print(cat)
````

Вне функции переменная `cat` не определена. 

Параметры тоже являются локальными.
Например, вне `cat_twice` не существует переменных `part1` и `part2`.

## Диаграммы стека

Чтобы отслеживать, какие переменные где можно использовать, иногда полезно рисовать **диаграмму стека**. 
Как и диаграммы состояний, диаграммы стека показывают значение каждой переменной, а также функцию, которой она принадлежит.

Каждая функция представлена **фреймом**.
Фрейм — это прямоугольник с именем функции снаружи и параметрами и локальными переменными внутри.

Вот диаграмма стека для предыдущего примера.

````python
from diagram import make_frame, Stack

d1 = dict(line1=line1, line2=line2)
frame1 = make_frame(d1, name='__main__', dy=-0.3, loc='left')

d2 = dict(part1=line1, part2=line2, cat=line1+line2)
frame2 = make_frame(d2, name='cat_twice', dy=-0.3, 
                    offsetx=0.03, loc='left')

d3 = dict(string=line1+line2)
frame3 = make_frame(d3, name='print_twice', 
                    offsetx=0.04, offsety=-0.3, loc='left')

d4 = {"?": line1+line2}
frame4 = make_frame(d4, name='print', 
                    offsetx=-0.22, offsety=0, loc='left')

stack = Stack([frame1, frame2, frame3, frame4], dy=-0.8)
````

````python
from diagram import diagram, adjust


width, height, x, y = [3.77, 2.9, 1.1, 2.65]
ax = diagram(width, height)
bbox = stack.draw(ax, x, y)
# adjust(x, y, bbox)

import matplotlib.pyplot as plt
plt.savefig('chap03_stack_diagram.png', dpi=300)
````

Фреймы располагаются в стеке, который показывает, какая функция вызвала
какую, и так далее. Снизу видно, что `print` был вызван `print_twice`, который вызвал `cat_twice`, который был вызван `__main__` — это специальное имя для верхнего фрейма.
Когда вы создаёте переменную вне какой-либо функции, она принадлежит `__main__`.

В фрейме для `print` вопросительный знак означает, что мы не знаем имя параметра.

## Обратные трассировки

Когда во время выполнения возникает ошибка, Python показывает имя функции, которая работала, имя функции, которая её вызвала, и так далее вверх по стеку.
Чтобы увидеть пример, я определю версию `print_twice`, содержащую ошибку — она пытается вывести `cat`, который является локальной переменной в другой функции.

````python
def print_twice(string):
    print(cat)            # NameError
    print(cat)
````

Теперь посмотрим, что произойдёт при запуске `cat_twice`.

````python
# This cell tells Jupyter to provide detailed debugging information
# when a runtime error occurs, including a traceback.

%xmode Verbose
````

````python
%%expect NameError

cat_twice(line1, line2)
````

Сообщение об ошибке содержит **traceback** — список функций, которые выполнялись в момент ошибки, и тех, что их вызывали.
В этом примере видно, что `cat_twice` вызвала `print_twice`, и ошибка возникла в `print_twice`.

Порядок функций в traceback совпадает с порядком фреймов на диаграмме стека.
Функция, выполнявшаяся во время ошибки, находится внизу.

## Зачем нужны функции?

Пока может быть не очевидно, зачем делить программу на функции.
На это есть несколько причин:

-   Создание новой функции даёт возможность дать имя группе инструкций,
    что делает программу легче читаемой и отлаживаемой.

-   Функции позволяют сократить программу, устранив повторяющийся код.
    Если позже что‑то изменится, исправление придётся сделать только в одном месте.

-   Разделив длинную программу на функции, можно отлаживать части поодиночке,
    а затем собрать их в рабочее целое.

-   Хорошо спроектированные функции часто полезны во многих программах. Однажды написав и отладив такую функцию, её можно переиспользовать.

## Отладка

Отладка может раздражать, но она также сложна, интересна и порой даже увлекательна.
Кроме того, это одно из самых важных умений, которое вы можете освоить.

В некотором смысле отладка похожа на детективную работу.
У вас есть подсказки, и нужно восстановить события, которые привели к наблюдаемым результатам.

Отладка также похожа на экспериментальную науку.
Когда у вас появляется предположение о том, что идёт не так, вы меняете программу и пробуете снова.
Если гипотеза верна, можно предсказать результат изменения и на шаг приблизиться к работающей программе.
Если гипотеза оказалась неверной, нужно придумать новую.

Для некоторых людей программирование и отладка — одно и то же; то есть программирование — это процесс постепенной отладки программы, пока она не начнёт делать то, что вам нужно.
Идея в том, что следует начать с работающей программы и вносить небольшие изменения, отлаживая их по ходу дела.

Если вы проводите много времени за отладкой, часто это означает, что вы пишете слишком много кода перед тем, как начать тесты.
Делая более мелкие шаги, вы можете двигаться быстрее.

## Глоссарий

**определение функции:**
Инструкция, создающая функцию.

**заголовок:**
 Первая строка определения функции.

**тело:**
 Последовательность инструкций внутри определения функции.

**объект функции:**
Значение, создаваемое определением функции.
Имя функции — это переменная, которая ссылается на объект функции.

**параметр:**
 Имя, используемое внутри функции для обращения к значению, переданному как аргумент.

**цикл:**
 Инструкция, которая выполняет одну или несколько других инструкций, часто многократно.

**локальная переменная:**
Переменная, определённая внутри функции и доступная только в ней.

**диаграмма стека:**
Графическое представление стека функций, их переменных и значений, на которые они ссылаются.

**фрейм:**
 Прямоугольник на диаграмме стека, представляющий вызов функции.
 Он содержит локальные переменные и параметры функции.

**трассировка (traceback):**
 Список выполняющихся функций, выводимый при возникновении исключения.
