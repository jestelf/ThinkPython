## Сравнение карт

Предположим, мы создаем второй объект `card` с тем же костюмом и рангом.

```python
queen2 = Card(1, 12)
print(queen2)
```

Если мы используем оператор `==` для их сравнения, он проверяет, относятся ли «Queen» и `Queen2` к одному и тому же объекту.

```python
queen == queen2
```

Они этого не делают, поэтому он возвращает `false '.
Мы можем изменить это поведение, определив специальный метод `__eq__`.

```python
%%add_method_to Card

    def __eq__(self, other):
        return self.suit == other.suit and self.rank == other.rank
```

`__eq__` принимает два объекта` card` в виде параметров и возвращает `true`, если они имеют одинаковый костюм и ранги, даже если они не один и тот же объект.
Другими словами, это проверяет, являются ли они эквивалентными, даже если они не идентичны.

Когда мы используем оператор `==` с объектами `card`, Python вызывает метод` __eq__`.

```python
queen == queen2
```

В качестве второго теста давайте создадим карту с одним и тем же костюмом и другим рангом.

```python
six = Card(1, 6)
print(six)
```

Мы можем подтвердить, что «Queen» и «Six» не эквивалентны.

```python
queen == six
```

Если мы используем оператор `! =`, Python вызывает специальный метод под названием `__ne__`, если он существует.
В противном случае он вызывает `__eq__` и инвертирует результат - поэтому, если `__eq__` возвращает` true`, результат оператора `! =` `False`.

```python
queen != queen2
```

```python
queen != six
```

Теперь предположим, что мы хотим сравнить две карты, чтобы увидеть, что больше.
Если мы используем одного из реляционных операторов, мы получим `typeError.

```python
%%expect TypeError

queen < queen2
```

Чтобы изменить поведение оператора «<», мы можем определить специальный метод под названием «__LT__`, который коротко для« меньше, чем ».
Ради этого примера давайте предположим, что костюм более важен, чем ранга, - поэтому все пики опережают все сердца, которые опережают все бриллианты и так далее.
Если две карты имеют одинаковый костюм, то, что с более высоким рангом выигрывает.

Чтобы реализовать эту логику, мы будем использовать следующий метод, который в этом порядке возвращает кортеж, содержащий костюм и ранги карты.

```python
%%add_method_to Card

    def to_tuple(self):
        return (self.suit, self.rank)
```

Мы можем использовать этот метод для написания `__lt__`.

```python
%%add_method_to Card

    def __lt__(self, other):
        return self.to_tuple() < other.to_tuple()
```

Сравнение кортежей сравнивает первые элементы из каждого кортежа, которые представляют костюмы.
Если они одинаковы, он сравнивает вторые элементы, которые представляют ряды.

Теперь, если мы используем оператор «<` », он вызывает метод` __LT__.

```python
six < queen
```

Если мы используем оператор `>`, он вызывает специальный метод под названием `__gt__`, если он существует.
В противном случае он вызывает `__lt__` с аргументами в противоположном порядке.

```python
queen < queen2
```

```python
queen > queen2
```

Наконец, если мы используем оператор `<=`, он вызывает специальный метод под названием `__le__`.

```python
%%add_method_to Card

    def __le__(self, other):
        return self.to_tuple() <= other.to_tuple()
```

Таким образом, мы можем проверить, меньше ли одна карта или равна другой.

```python
queen <= queen2
```

```python
queen <= six
```

Если мы используем оператор `> =`, он использует `__ge__`, если он существует. В противном случае он вызывает `__le__` с аргументами в противоположном порядке.

```python
queen >= six
```

Как мы их определили, эти методы полны в том смысле, что мы можем сравнить любые два объекта «карты» и последовательны в том смысле, что результаты разных операторов не противоречат друг другу.
С этими двумя свойствами мы можем сказать, что объекты `card` ** полностью заказаны **.
И это означает, как мы скоро увидим, что их можно отсортировать.

## палубы

Теперь, когда у нас есть объекты, которые представляют карты, давайте определим объекты, которые представляют колоды.
Ниже приведено определение класса для `deck` с
Метод `__init__ 'принимает список объектов` card` в качестве параметра и присваивает его атрибуту с именем `cards'.

```python
class Deck:

    def __init__(self, cards):
        self.cards = cards
```

Чтобы создать список, который содержит 52 карты в стандартной колоде, мы будем использовать следующий статический метод.

```python
%%add_method_to Deck

    def make_cards():
        cards = []
        for suit in range(4):
            for rank in range(2, 15):
                card = Card(suit, rank)
                cards.append(card)
        return cards
```

В `make_cards` внешний цикл перечисляет костюмы от` 0` до `3`.
Внутренняя петля перечисляет ряды от `2` до` 14` - где `14` представляет туз, который опередил короля.
Каждая итерация создает новую «карту» с текущим костюмом и рангом и добавляет ее на «карты».

Вот как мы составляем список карт и объект «колоды», который содержит его.

```python
cards = Deck.make_cards()
deck = Deck(cards)
len(deck.cards)
```

Он содержит 52 карты, как и предполагалось.

## родители и дети

Наследование - это способность определить новый класс, который является модифицированной версией существующего класса.
Например, допустим, мы хотим, чтобы класс представил «руку», то есть карты, удерживаемые одним игроком.

* Рука похожа на колоду - оба состоят из коллекции карт, и оба требуют таких операций, как добавление и удаление карт.

* Рука также отличается от колоды - есть операции, которые мы хотим для рук, которые не имеют смысла для колоды. Например, в покере мы можем сравнить две руки, чтобы увидеть, какая победа. В мосту мы могли бы вычислить счет за руку, чтобы сделать ставку.

Эта связь между классами, где одна является специализированной версией другой, поддается наследству. 

Чтобы определить новый класс, основанный на существующем классе, мы поместили имя существующего класса в скобках.

```python
class Hand(Deck):
    """Represents a hand of playing cards."""
```

Это определение указывает на то, что «рука» наследует от `deck`, что означает, что объекты` hand 'могут получить доступ к методам, определенным в «Deck», например, `take_card` и` put_card`.

`Hand 'также наследует` __init__` от `deck`, но если мы определим` __init__` в классе `hand`, он переопределяет тот в классе« Deck ».

```python
%%add_method_to Hand

    def __init__(self, label=''):
        self.label = label
        self.cards = []
```

Эта версия `__init__` принимает дополнительную строку в качестве параметра и всегда начинается с пустого списка карт.
Когда мы создаем «руку», Python вызывает этот метод, а не тот, который в «Deck» - который мы можем подтвердить, проверив, что результат имеет атрибут `label '.

```python
hand = Hand('player 1')
hand.label
```

Чтобы разобраться с картой, мы можем использовать `take_card`, чтобы удалить карту из` deck 'и `put_card`, чтобы добавить карту в` hand'.

```python
deck = Deck(cards)
card = deck.take_card()
hand.put_card(card)
print(hand)
```

Давайте инкапсулируем этот код в метод «Deck» с именем `move_cards`.

```python
%%add_method_to Deck

    def move_cards(self, other, num):
        for i in range(num):
            card = self.take_card()
            other.put_card(card)
```

Этот метод является полиморфным, то есть он работает с более чем одним типом: «Self» и «Другое» могут быть либо «рукой», либо «палубой».
Таким образом, мы можем использовать этот метод, чтобы сдать карту от «палубы» до `hand ', от одной` hand' до другой или от `hand 'обратно в` deck'.

Когда новый класс наследует от существующего, существующий называется ** родитель **, а новый класс называется ** ребенок **. В общем:

* Экземпляры детского класса должны иметь все атрибуты родительского класса, но они могут иметь дополнительные атрибуты.

* Дочерний класс должен иметь все методы родительского класса, но он может иметь дополнительные методы.

* Если дочерний класс переопределяет метод из родительского класса, новый метод должен принимать одинаковые параметры и вернуть совместимый результат.

Этот набор правил называется «Принцип замены Лискова» после того, как компьютерный ученый Барбара Лисков.

Если вы следуете этим правилам, любая функция или метод, предназначенные для работы с экземпляром родительского класса, например, «колода», также будут работать с экземплярами детского класса, например, «рука».
Если вы нарушаете эти правила, ваш код будет рухнуть, как карточный дом (извините).

## Отладка наследования является полезной функцией. Некоторые программы, которые были бы повторяющимися без наследства, могут быть написаны более кратко с ним. Кроме того, наследование может облегчить повторное использование кода, поскольку вы можете настроить поведение родительского класса без необходимости его изменений. В некоторых случаях структура наследования отражает естественную структуру проблемы, которая облегчает понимание дизайна.  С другой стороны, наследство может затруднить чтение программ. Когда метод вызывается, иногда неясно, где найти его определение - соответствующий код может распространяться по нескольким модулям.  Каждый раз, когда вы не уверены в потоке выполнения через вашу программу, самое простое решение - добавить печатные операторы в начале соответствующих методов. Если `deck.shuffle` напечатает сообщение, в котором говорится что -то вроде` running deck.shuffle ', то, когда программа запускает, она прослеживает поток выполнения.  В качестве альтернативы вы можете использовать следующую функцию, которая принимает
Объект и имя метода (как строка) и возвращает класс, который предоставляет определение метода.

```python
def find_defining_class(obj, method_name):
    """Find the class where the given method is defined."""
    for typ in type(obj).mro():
        if method_name in vars(typ):
            return typ
    return f'Method {method_name} not found.'
```

`find_defining_class` использует метод` mro`, чтобы получить список объектов класса (типы), которые будут искать методы.
«MRO» означает «Порядок разрешения метода», который представляет собой последовательность классов Python Searchs, чтобы «разрешить» имя метода, то есть найти объект функции, на который ссылается имя, на которое ссылается.

В качестве примера давайте создам создание «Бриджханд», а затем найдем определяющий класс «shuffle».

```python
hand = BridgeHand('player 3')
find_defining_class(hand, 'shuffle')
```

Метод `shuffle` для объекта« Бриджханд » - это то, что в« колоде ».

Следующие несколько упражнений просят вам написать функции, которые классифицируют покерные руки.
Если вы не знакомы с покером, я объясню, что вам нужно знать.
Мы будем использовать следующий класс, чтобы представлять покерные руки.

```python
class PokerHand(Hand):
    """Represents a poker hand."""

    def get_suit_counts(self):
        counter = {}
        for card in self.cards:
            key = card.suit
            counter[key] = counter.get(key, 0) + 1
        return counter
    
    def get_rank_counts(self):
        counter = {}
        for card in self.cards:
            key = card.rank
            counter[key] = counter.get(key, 0) + 1
        return counter    
```

`PokerHand 'предоставляет два метода, которые помогут с упражнениями.

* `get_suit_counts` щедет через карты в` pokerhand ', подсчитывает количество карт в каждом костюме и возвращает словарь, который отображает из каждого кода иска в количество раз.

* `get_rank_counts` делает то же самое с рядами карт, возвращая словарь, который отображает из каждого кода ранга в количество раз.

Это упражнение является предостерегающей историей об общей ошибке, которая может быть трудно отлаживать.
Рассмотрим следующее определение класса.

```python
class Kangaroo:
    """A Kangaroo is a marsupial."""
    
    def __init__(self, name, contents=[]):
        """Initialize the pouch contents.

        name: string
        contents: initial pouch contents.
        """
        self.name = name
        self.contents = contents

    def __str__(self):
        """Return a string representaion of this Kangaroo.
        """
        t = [ self.name + ' has pouch contents:' ]
        for obj in self.contents:
            s = '    ' + object.__str__(obj)
            t.append(s)
        return '\n'.join(t)

    def put_in_pouch(self, item):
        """Adds a new item to the pouch contents.

        item: object to be added
        """
        self.contents.append(item)
```

`__init__` принимает два параметра:« Имя

`__str__` возвращает строковое представление объекта, который включает имя и содержимое мешочка.

`put_in_pouch` принимает любой объект и добавляет его на` contents`.

Теперь посмотрим, как работает этот класс.
Мы создадим два объекта `kangaroo 'с именами`' kanga' 'и `' roo '.
