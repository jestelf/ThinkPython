## Аргументы

При вызове функции выражение в скобках называется * * аргументом**.
Обычно я объясняю, почему, но в этом случае техническое значение термина почти не имеет ничего общего с общим значением этого слова, поэтому я даже не буду пытаться.

Некоторые из функций, которые мы видели до сих пор, принимают только один аргумент, например, «int».

````python
int('101')
````

Некоторые берут два, например «math.pow».

````python
math.pow(5, 2)
````

Некоторые могут принимать дополнительные аргументы, которые являются необязательными. 
Например, `int` может принимать второй аргумент, указывающий основание числа.

````python
int('101', 2)
````

Последовательность цифр «101» в основании 2 представляет собой число 5 в основании 10.

`round` также принимает необязательный второй аргумент, который является числом знаков после запятой, до которого нужно округлить.

````python
round(math.pi, 3)
````

Некоторые функции могут принимать любое количество аргументов, например `print`.

````python
print('Any', 'number', 'of', 'arguments')
````

Если вы вызываете функцию и предоставляете слишком много аргументов, это `TypeError`.

````python
%%expect TypeError

float('123.0', 2)
````

Если вы предоставляете слишком мало аргументов, это также `TypeError`.

````python
%%expect TypeError

math.pow(2)
````

И если вы предоставляете аргумент с типом, который функция не может обработать, это тоже `TypeError`.

````python
%%expect TypeError

math.sqrt('123')
````

Такая проверка может раздражать, когда вы начинаете, но она помогает обнаруживать и исправлять ошибки.

## Комментарии

По мере того, как программы становятся больше и сложнее, их становится труднее читать.
Формальные языки плотные, и часто бывает трудно взглянуть на кусок кода и понять, что он делает и почему.

По этой причине рекомендуется добавлять заметки к своим программам, чтобы объяснить на естественном языке, что делает программа. 
Эти заметки называются * * комментариями * * и начинаются с символа #.

````python
# number of seconds in 42:42
seconds = 42 * 60 + 42
````

В этом случае комментарий появляется на строке сам по себе. Вы также можете поставить
комментарии в конце строки:

````python
miles = 10 / 1.61     # 10 kilometers in miles
````

Все, от «#» до конца строки, игнорируется --- у него нет
влияние на выполнение программы.

Комментарии наиболее полезны, когда они документируют неочевидные особенности кода.
Разумно предположить, что читатель может понять *что* делает код; полезнее объяснить *почему*.

Этот комментарий лишний с кодом и бесполезный:

````python
v = 8     # assign 8 to v
````

Этот комментарий содержит полезную информацию, которой нет в коде:

````python
v = 8     # velocity in miles per hour 
````

Хорошие имена переменных могут уменьшить потребность в комментариях, но длинные имена могут
сделать сложные выражения трудными для чтения, поэтому есть компромисс.

## Отладка

В программе могут возникать три вида ошибок: синтаксические, ошибки времени выполнения и семантические ошибки. Полезно различать их, чтобы быстрее находить причину проблемы.

* **Синтаксическая ошибка**: «Синтаксис» относится к структуре программы и правилам этой структуры. Если где‑то в программе есть синтаксическая ошибка, Python не запустит её и сразу покажет сообщение об ошибке.

* **Ошибка времени выполнения**: если синтаксических ошибок нет, программа может начать выполняться. Но если что‑то пойдет не так, Python выведет сообщение об ошибке и остановится. Такой тип ошибок называется ошибкой времени выполнения, или **исключением**, потому что говорит о какой‑то необычной ситуации.

* **Семантическая ошибка**: третий тип ошибок — «семантические», то есть связанные со смыслом. Если в программе есть семантическая ошибка, она выполняется без сообщений об ошибке, но делает не то, что вы ожидали. Найти такие ошибки бывает сложно: нужно проанализировать результаты работы программы и попытаться понять, что происходит.

Как мы видели, недопустимое имя переменной является синтаксической ошибкой.

````python
%%expect SyntaxError

million! = 1000000
````

Если вы используете оператор с типом, который он не поддерживает, это ошибка времени выполнения.

````python
%%expect TypeError

'126' / 3
````

Наконец, вот пример семантической ошибки.
Предположим, мы хотим вычислить среднее значение `1` и `3`, но забываем о порядке операций и пишем так:

````python
1 + 3 / 2
````

Когда это выражение вычисляется, оно не выдает сообщение об ошибке, поэтому нет синтаксической ошибки или ошибки во время выполнения.
Но результат не является средним из «1» и «3», поэтому программа неверна.
Это семантическая ошибка, потому что программа работает, но не делает то, что задумано.

## Глоссарий

**переменная:**
Имя, которое ссылается на значение.

**оператор присваивания:**
Инструкция, которая присваивает переменной значение.

**диаграмма состояния:**
Графическое представление набора переменных и значений, на которые они указывают.

**ключевое слово:**
Специальное слово, которое задает структуру программы.

**инструкция import:**
Инструкция, читающая файл модуля, чтобы мы могли использовать содержащиеся в нем переменные и функции.

**модуль:**
Файл, содержащий код на Python, включая определения функций и иногда другие инструкции.

**оператор точки:**
Оператор `.`, используемый для доступа к функции в другом модуле: указывается имя модуля, точка и имя функции.

**вычислить:**
Выполнить операции в выражении, чтобы получить значение.

**инструкция:**
Одна или несколько строк кода, представляющих команду или действие.

**выполнить:**
Запустить инструкцию и сделать то, что в ней сказано.

**аргумент:**
Значение, передаваемое функции при её вызове.

**комментарий:**
Текст в программе, поясняющий её работу, но не влияющий на выполнение.

**ошибка времени выполнения:**
Ошибка, из-за которой программа выводит сообщение об ошибке и завершается.

**исключение:**
Ошибка, обнаруженная во время работы программы.

**семантическая ошибка:**
Ошибка, из-за которой программа делает не то, что нужно, но не выводит сообщение об ошибке.
