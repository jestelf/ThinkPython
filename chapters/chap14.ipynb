## чистые функции

Мы можем использовать `copy` для написания чистых функций, которые не изменяют их параметры.
Например, вот функция, которая занимает объект «время» и продолжительность часа, минуты и секунды.
Он делает копию исходного объекта, использует `urment_time` для изменения копии и возвращает ее.


```python
def add_time(time, hours, minutes, seconds):
    total = copy(time)
    increment_time(total, hours, minutes, seconds)
    return total
```

Вот как мы его используем.


```python
end = add_time(start, 1, 32, 0)
print_time(end)
```

Возвращаемое значение - это новый объект, представляющий время окончания фильма.
И мы можем подтвердить, что `start 'не изменился.


```python
print_time(start)
```

`add_time` - это **чистая функция**, потому что она не изменяет ни один из объектов, передаваемых ему в качестве аргументов, и его единственный эффект - вернуть значение.

Все, что можно сделать с нечистыми функциями, также может быть сделано с помощью чистых функций.
На самом деле, некоторые языки программирования позволяют только чистые функции.
Программы, которые используют чистые функции, могут быть менее подверженными ошибкам, но нечистые функции иногда удобны и могут быть более эффективными.

В целом, я предлагаю вам написать чистые функции всякий раз, когда это разумно, и прибегать к нечистым функциям, только если есть убедительное преимущество.
Этот подход можно назвать**функциональным стилем программирования**.


## Прототип и патч

В предыдущем примере `increment_time` и` add_time`, кажется, работают, но если мы попробуем другой пример, мы увидим, что они не совсем правы.

Предположим, вы приходите в театр и обнаружите, что фильм начинается в 9: 40 `, а не« 9: 20 `.
Вот что происходит, когда мы вычисляем обновленное время окончания.


```python
start = make_time(9, 40, 0)
end = add_time(start, 1, 32, 0)
print_time(end)
```

Результат не является достоверным временем.
Проблема заключается в том, что `increment_time` не касается случаев, когда количество секунд или минут наносит более чем« 60 ».

Вот улучшенная версия, которая проверяет, превышает ли «второе» или равняется `60` - если да, то увеличивает` minute` - затем проверяет, превышает ли `minute` или равно` 60` - если это так, это увеличивает `час


```python
def increment_time(time, hours, minutes, seconds):
    time.hour += hours
    time.minute += minutes
    time.second += seconds

    if time.second >= 60:
        time.second -= 60
        time.minute += 1

    if time.minute >= 60:
        time.minute -= 60
        time.hour += 1
```

Исправление `increment_time` также исправляет` add_time`, который использует его.
Так что теперь предыдущий пример работает правильно.


```python
end = add_time(start, 1, 32, 0)
print_time(end)
```

Но эта функция все еще не верна, потому что аргументы могут быть больше, чем `60 '.
Например, предположим, что нам дают время выполнения как «92» минуты, а не «1» и «32» минуты.
Мы могли бы назвать «add_time».


```python
end = add_time(start, 0, 92, 0)
print_time(end)
```

Результат не является достоверным временем.
Итак, давайте попробуем другой подход, используя функцию «divmod».
Мы сделаем копию «Start» и изменим ее, увеличив атрибут `Minute '.


```python
end = copy(start)
end.minute = start.minute + 92
end.minute
```

Теперь `minute` -` 132`, что составляет `2` часы и `12 'минут.
Мы можем использовать «divmod», чтобы разделить на `60 'и вернуть количество целых часов и количество оставшихся минут.


```python
carry, end.minute = divmod(end.minute, 60)
carry, end.minute
```

Теперь «минута» верна, и мы можем добавить часы к «часу».


```python
end.hour += carry
print_time(end)
```

Результат - достоверное время.
Мы можем сделать то же самое с «Часом» и «вторым» и инкапсулировать весь процесс в функцию.


```python
def increment_time(time, hours, minutes, seconds):
    time.hour += hours
    time.minute += minutes
    time.second += seconds
    
    carry, time.second = divmod(time.second, 60)
    carry, time.minute = divmod(time.minute + carry, 60)
    carry, time.hour = divmod(time.hour + carry, 24)
```

С этой версией `urment_time`` add_time` работает правильно, даже если аргументы превышают `60 '.


```python
end = add_time(start, 0, 90, 120)
print_time(end)
```

Этот раздел демонстрирует план разработки программы, который я называю **прототипом и патч**.
Мы начали с простого прототипа, который работал правильно для первого примера.
Затем мы проверили его с более сложными примерами - когда мы нашли ошибку, мы изменили программу, чтобы исправить ее, например, нанести патч на шину с прокол.

Этот подход может быть эффективным, особенно если у вас еще нет глубокого понимания проблемы.
Но постепенные исправления могут генерировать код, который излишне сложный - поскольку он имеет дело со многими особыми случаями - и ненадежным - поскольку трудно понять, есть ли у вас
Нашел все ошибки.


## Design-Pirst Development

Альтернативным планом является **проектирование-первое развитие**, которое включает в себя больше планирования перед прототипированием. В процессе первого дизайна иногда высокое понимание проблемы облегчает программирование.

В этом случае понимание состоит в том, что мы можем думать о объекте «Время» как о трехзначном числе в базе 60, также известном как сексмальный.
Атрибут «второй» - это «столбец», атрибут `Minute` -« столбец шестидесятых »,
А атрибут «часа»-это «тридцать шесть сотни сотни столбцов».
Когда мы написали `urment_time`, мы эффективно делали дополнение в базе 60, поэтому мы должны были перенести из одного столбца в другой.

Это наблюдение предлагает другой подход ко всей проблеме - мы можем преобразовать объекты «время» в целые числа и воспользоваться тем фактом, что Python знает, как выполнять целочисленную арифметику.

Вот функция, которая преобразуется из «времени» в целое число.


```python
def time_to_int(time):
    minutes = time.hour * 60 + time.minute
    seconds = minutes * 60 + time.second
    return seconds
```

Результатом является количество секунд с начала дня.
Например, `01: 01: 01` - 1` час, `1` минута и` 1` секунды с начала дня, которая является суммой `3600` секунды, `60` 1 секунды и `1` второй.


```python
time = make_time(1, 1, 1)
print_time(time)
time_to_int(time)
```

А вот функция, которая идет в другом направлении - преобразование целого числа в объект «время» - с использованием функции «divmod».


```python
def int_to_time(seconds):
    minute, second = divmod(seconds, 60)
    hour, minute = divmod(minute, 60)
    return make_time(hour, minute, second)
```

Мы можем проверить его, преобразив предыдущий пример обратно в «время».


```python
time = int_to_time(3661)
print_time(time)
```

Используя эти функции, мы можем написать более краткую версию `add_time`.


```python
def add_time(time, hours, minutes, seconds):
    duration = make_time(hours, minutes, seconds)
    seconds = time_to_int(time) + time_to_int(duration)
    return int_to_time(seconds)
```

Первая строка преобразует аргументы в объект «время», называемый «продолжительность».
Вторая строка преобразует «время» и «продолжительность» в секунды и добавляет их.
Третья строка преобразует сумму в объект «время» и возвращает ее.

Вот как это работает.


```python
start = make_time(9, 40, 0)
end = add_time(start, 1, 32, 0)
print_time(end)
```

В некотором смысле преобразование из базы 60 в основание 10 и назад сложнее, чем
Просто дело со временем. Базовое преобразование более абстрактно; наша интуиция
Для работы с временными значениями лучше.

Но если у нас есть понимание, чтобы рассматривать время как базовые 60 чисел - и инвестировать усилия, чтобы написать функции преобразования `time_to_int` и` int_to_time` - мы получаем программу, которая легче, проще для чтения и отладка и более надежных.

Также легче добавить функции позже. Например, представьте, что вы вычитаете два объекта «время», чтобы найти продолжительность между ними.
Наивный подход заключается в реализации вычитания с заимствованием.
Использование функций преобразования проще и с большей вероятностью будет правильным.

По иронии судьбы, иногда усложняет проблему - или более общий - облегчает, потому что существует меньше особых случаев и меньше возможностей для ошибок.
