Некоторые из упражнений в этой книге могут быть выполнены кратко и эффективно с наборами.

```python
def has_duplicates(t):
    d = {}
    for x in t:
        d[x] = True
    return len(d) < len(t)
```

```python
def has_duplicates(t):
    s = set(t)
    return len(s) < len(t)
```

Отношения между множествами проверяются при помощи операторов сравнения.
Например, оператор `<=` проверяет, является ли одно множество подмножеством другого, включая случай равенства.

С помощью этих операторов можно решить некоторые упражнения из главы 7. Например, ниже показана версия функции `uses_only`, использующая цикл.

```python
def uses_only(word, available):
    for letter in word: 
        if letter not in available:
            return False
    return True
```

Функция `uses_only` проверяет, входят ли все буквы из `word` в `available`.
С использованием множеств её можно переписать так.

```python
def uses_only(word, available):
    return set(word) <= set(available)
```

Если буквы из `word` образуют подмножество букв в `available`, это означает, что `word` использует только буквы из `available`.

В одном из упражнений в [Глава 11] (Глава_TUPLES) я сделал словарь, который отображает из сортированной строки букв в список слов, которые можно писать этими буквами.
Например, строка «opst» карты в список `['opts', 'post', 'pots', 'spot', 'stop', 'tops']`.
Вот оригинальный код.

```python
def all_anagrams(filename):
    d = {}
    for line in open(filename):
        word = line.strip().lower()
        t = signature(word)
        if t not in d:
            d[t] = [word]
        else:
            d[t].append(word)
    return d
```

А вот более простая версия с использованием `defaultDict '.

```python
def all_anagrams(filename):
    d = defaultdict(list)
    for line in open(filename):
        word = line.strip().lower()
        t = signature(word)
        d[t].append(word)
    return d
```

## условные выражения

Условные утверждения часто используются для выбора одного из двух значений, например, это:

```python
import math
x = 5
```

```python
if x > 0:
    y = math.log(x)
else:
    y = float('nan')
```

```python
y
```

Это утверждение проверяет, является ли `x` положительным.Если так, то он вычисляет его логарифм.
Если нет, то `math.log` поднимет стоимость.
Чтобы избежать остановки программы, мы генерируем «NAN», который является специальным значением с плавающей запятой, которая представляет «не число».

Мы можем написать это утверждение более кратко, используя ** условное выражение **.

```python
y = math.log(x) if x > 0 else float('nan')
```

```python
y
```

Вы можете почти прочитать эту строку, как английский: «y` получает log- `x`, если` x` больше 0; в противном случае он получает `nan`».

Рекурсивные функции иногда можно писать кратко с использованием условных выражений.
Например, вот версия `factorial` с условным _statement_.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

А вот версия с условной _expression_.

```python
def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)
```

Другим использованием условных выражений является обработка дополнительных аргументов.
Например, вот определение класса с методом `__init__`, который использует условное операцию для проверки параметра со значением по умолчанию.

```python
class Kangaroo:
    def __init__(self, name, contents=None):
        self.name = name
        if contents is None:
            contents = []
        self.contents = contents
```

Вот версия, которая использует условное выражение.

```python
def __init__(self, name, contents=None):
    self.name = name
    self.contents = [] if contents is None else contents 
```

В целом, вы можете заменить условное утверждение на условное выражение, если обе ветви содержат одно выражение и отсутствие операторов.

В предыдущих главах мы видели несколько примеров, когда мы начинаем с пустого списка и добавляем элементы по одному, используя метод «Приложение».

## `any 'и` all

Python обеспечивает встроенную функцию, «любой», которая принимает последовательность логических значений и возвращает `true`, если какое-либо из значений является« true ».

```python
any([False, False, True])
```

«Любой» часто используется с выражениями генератора.

```python
any(letter == 't' for letter in 'monty')
```

Этот пример не очень полезен, потому что он делает то же самое, что и оператор «в».
Но мы могли бы использовать «любое», чтобы написать краткие решения для некоторых упражнений в [Глава 7] (глава_SEARCH).Например, мы можем написать `change_none` как это.

```python
def uses_none(word, forbidden):
    """Checks whether a word avoids forbidden letters."""
    return not any(letter in forbidden for letter in word)
```

```python
uses_none('banana', 'xyz')
```

```python
uses_none('apple', 'efg')
```

Эта функция проходит через буквы в «Word» и проверяет, находятся ли какие -либо из них в «запрещенном».
Использование «любого» с выражением генератора является эффективным, потому что оно немедленно останавливается, если оно находит значение «истинного», поэтому он не должен проходить через всю последовательность.

Python предоставляет еще одну встроенную функцию, «All», которая возвращает `true`, если каждый элемент последовательности` true '.
Мы можем использовать его, чтобы написать краткую версию `using_all`.

```python
def uses_all(word, required):
    """Check whether a word uses all required letters."""
    return all(letter in word for letter in required)
```

```python
uses_all('banana', 'ban')
```

```python
uses_all('apple', 'api')
```

Выражения с использованием «any» и «all» могут быть кратким, эффективным и простым в чтении.

```python
class Point:
    """Represents a point in 2-D space."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return f'({self.x}, {self.y})'
```

Это много кода, чтобы передать небольшое количество информации.
`nameTuple` обеспечивает более краткий способ определить такие занятия.

```python
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
```

Первый аргумент - это имя класса, который вы хотите создать.А
Во -вторых, это список объектов атрибутов `point`, которые должны иметь.
Результатом является объект класса, поэтому он назначен капитализированным именем переменной.

Класс, созданный с помощью `nameTuple`, предоставляет метод` __init__`, который присваивает значения атрибутам и «__str__`, который отображает объект в читаемой форме.
Таким образом, мы можем создавать и отобразить такой объект «точка».

```python
p = Point(1, 2)
p
```

`Point` также предоставляет метод` __eq__`, который проверяет, эквивалентны ли два объекта «точка», то есть, являются ли их атрибуты одинаковыми.

```python
p == Point(1, 2)
```

## Упаковывать аргументы ключевого слова
В [Глава 11] (пакет аргументов в разделе) мы пишем функцию, которая упаковывает ее аргументы в кортеж.

```python
def mean(*args):
    return sum(args) / len(args)
```

Вы можете вызвать эту функцию с любым количеством аргументов.

```python
mean(1, 2, 3)
```

Но оператор `*` не упаковывает аргументы ключевых слов.
Таким образом, вызов этой функции с помощью аргумента ключевого слова вызывает ошибку.

```python
%%expect TypeError

mean(1, 2, start=3)
```

Чтобы упаковать аргументы ключевого слова, мы можем использовать оператор `**`:

```python
def mean(*args, **kwargs):
    print(kwargs)
    return sum(args) / len(args)
```

Параметр по составлению ключевых слов может иметь любое имя, но `kwargs 'является общим выбором.
Результатом является словарь, который отображает от ключевых слов к значениям.

```python
mean(1, 2, start=3)
```

В этом примере значение «kwargs» напечатано, но в противном случае не имеет никакого эффекта.

Но оператор `** также может быть использован в списке аргументов для распаковки словаря.
Например, вот версия `ead`, в которой есть какие -либо аргументы ключевых слов, которые он получает, а затем распаковывает их как аргументы ключевых слов для` sum '.

```python
def mean(*args, **kwargs):
    return sum(args, **kwargs) / len(args)
```

Теперь, если мы назовут «ead» с «start» в качестве аргумента ключевого слова, он передается на сумму, который использует его в качестве отправной точки суммирования.
В следующем примере `start = 3` добавляет` 3` к сумме перед вычислением среднего, поэтому сумма равен `6`, а результат -` 3`.

```python
mean(1, 2, start=3)
```

В качестве другого примера, если у нас есть словарь с клавишами `x` и` y`, мы можем использовать его с оператором распаковки для создания объекта «точка».

```python
d = dict(x=1, y=2)
Point(**d)
```

Без оператора распаков, «D` рассматривается как единый позиционный аргумент, поэтому он назначается« X », и мы получаем« typeerror », потому что нет второго аргумента, чтобы назначить` y`.

```python
%%expect TypeError

d = dict(x=1, y=2)
Point(d)
```

Когда вы работаете с функциями, которые имеют большое количество аргументов ключевых слов, часто полезно создавать и разобраться в словари, которые указывают часто используемые параметры.

```python
def pack_and_print(**kwargs):
    print(kwargs)
    
pack_and_print(a=1, b=2)
```

## Отладка

В предыдущих главах мы использовали `doctest` для тестирования функций.
Например, функция `add` принимает два числа и возвращает их сумму.
В её строке документа есть doctest, проверяющий, что `2 + 2` равно `4`.

```python
def add(a, b):
    '''Add two numbers.
    
    >>> add(2, 2)
    4
    '''
    return a + b
```

Эта функция принимает объект функции и запускает её doctest-тесты.

```python
from doctest import run_docstring_examples

def run_doctests(func):
    run_docstring_examples(func, globals(), name=func.__name__)
```

Поэтому протестировать `add` можно так.

```python
run_doctests(add)
```

Выходных данных нет, что означает успешное прохождение всех тестов.

В Python есть ещё один инструмент для автоматизированного тестирования — `unittest`.
Пользоваться им немного сложнее, но вот пример.

```python
from unittest import TestCase

class TestExample(TestCase):

    def test_add(self):
        result = add(2, 2)
        self.assertEqual(result, 4)
```

Сначала мы импортируем `TestCase` — класс из модуля `unittest`.
Чтобы воспользоваться им, нужно определить новый класс, наследующий `TestCase`, и добавить хотя бы один метод теста.
Имя метода теста должно начинаться с `test` и отражать, какую функцию он проверяет.

В нашем примере `test_add` вызывает `add`, сохраняет результат и использует унаследованный от `TestCase` метод `assertEqual`.
`assertEqual` принимает два аргумента и проверяет, равны ли они.

Чтобы запустить этот тест, нужно вызвать функцию `main` из `unittest`, передав ей несколько именованных аргументов.
Следующая функция показывает детали — если интересно, попросите виртуального помощника объяснить, как она работает.

```python
import unittest

def run_unittest():
    unittest.main(argv=[''], verbosity=0, exit=False)
```

`run_unittest` не принимает `TestExample` как аргумент — вместо этого она ищет классы, наследующие `TestCase`.
Затем она находит методы, начинающиеся с `test`, и выполняет их.
Этот процесс называется **обнаружением тестов**.

Посмотрим, что произойдёт при вызове `run_unittest`.

```python
run_unittest()
```

`unittest.main` сообщает, сколько тестов было выполнено, и их результат.
Здесь `OK` означает, что все тесты прошли.

Чтобы увидеть, что бывает при сбое теста, добавим в `TestExample` ошибочный метод.

```python
%%add_method_to TestExample

    def test_add_broken(self):
        result = add(2, 2)
        self.assertEqual(result, 100)
```

Вот что происходит при запуске тестов.

```python
run_unittest()
```

В отчёте указано, какой тестовый метод провалился, и приведено сообщение об ошибке.
В сводке указано, что было запущено два теста и один из них не прошёл.
