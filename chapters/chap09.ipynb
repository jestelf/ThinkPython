## списки изменяются

Чтобы прочитать элемент списка, мы можем использовать оператор кронштейна.
Индекс первого элемента - `0`.

```python
cheeses[0]
```

В отличие от струн, списки являются изменчивыми. Когда оператор кронштейна появляется на
Левая сторона назначения, он идентифицирует элемент списка
это будет назначено.

```python
numbers[1] = 17
numbers
```

Второй элемент «чисел», который раньше был `123`, теперь` 17 '.

Индексы списка работают так же, как строковые индексы:

- Любое целочисленное выражение может использоваться в качестве индекса.

- Если вы попытаетесь прочитать или написать элемент, которого не существует, вы получите
    `indexerror '.

- Если индекс имеет отрицательное значение, он считается отсталым от конца
    Список.

Оператор `in` работает в списках - он проверяет, появляется ли данный элемент где -нибудь в списке.

```python
'Edam' in cheeses
```

```python
'Wensleydale' in cheeses
```

## псевдоним

Если `a` относится к объекту, и вы назначаете` b = a`, то обе переменные относятся к одному и тому же объекту.

```python
a = [1, 2, 3]
b = a
b is a
```

Так что диаграмма состояния выглядит так.

```python
t = [1, 2, 3]
binding1 = Binding(Value('a'), Value(repr(t)), dy=-0.11)
binding2 = Binding(Value('b'), draw_value=False, dy=0.11)
frame = Frame([binding1, binding2], dy=-0.25)
```

```python
width, height, x, y = [1.11, 0.81, 0.17, 0.56]
ax = diagram(width, height)
bbox = frame.draw(ax, x, y)
# adjust(x, y, bbox)
```

Ассоциация переменной с объектом называется ** ссылка **.
В этом примере есть две ссылки на один и тот же объект.

У объекта с более чем одной ссылкой есть более одного имени, поэтому мы говорим, что объект ** псевдоним **.
Если псевдоним, объект, изменен, изменения, внесенные с одним именем, влияют на другое.
В этом примере, если мы изменим объект `B` относится, мы также меняем объект` a` относится.

```python
b[0] = 5
a
```

Итак, мы бы сказали, что «a» видит »это изменение.
Хотя это поведение может быть полезным, оно подвержено ошибкам.
В целом, безопаснее избегать псевдонима, когда вы работаете с изменчивыми объектами.

Для неизменных объектов, таких как строки, псевдоним не такая проблема.
В этом примере:

```python
a = 'banana'
b = 'banana'
```

Это почти никогда не имеет значения, ссылаются ли `a` и` b
строка или нет.

## Создание списка слов

В предыдущей главе мы читаем файл `words.txt` и искали слова с определенными свойствами, например, используя букву` e`.
Но мы читаем весь файл много раз, что не эффективно.
Лучше прочитать файл один раз и поместить слова в список.
Следующий цикл показывает, как.

```python
download('https://raw.githubusercontent.com/AllenDowney/ThinkPython/v3/words.txt');
```

```python
word_list = []

for line in open('words.txt'):
    word = line.strip()
    word_list.append(word)
    
len(word_list)
```

Перед циклом `word_list` инициализируется с пустым списком.
Каждый раз через цикл метод «Приложение» добавляет слово к концу.

Еще один способ сделать то же самое - использовать `read`, чтобы прочитать весь файл в строку.

```python
string = open('words.txt').read()
len(string)
```

Результатом является единственная строка с более чем миллионами символов.
Мы можем использовать метод «разделения», чтобы разделить его на список слов.

```python
word_list = string.split()
len(word_list)
```

Теперь, чтобы проверить, появляется ли в списке строку, мы можем использовать оператор `in`.

## отладка

Обратите внимание, что большинство методов списка изменяют аргумент и возвращают «нет».
Это противоположность методам струн, которые возвращают новую строку и оставляют оригинал в покое.

Если вы привыкли к написанию строкового кода, как это:

```python
word = 'plumage!'
word = word.strip('!')
word
```

Заманчиво написать код списка, как это:

```python
t = [1, 2, 3]
t = t.remove(3)           # WRONG!
```

`Удалить 'изменяет список и возвращает« нет », поэтому следующая операция, которую вы выполняете с помощью` t`, скорее всего, потерпит неудачу.

```python
%%expect AttributeError

t.remove(2)
```

Это сообщение об ошибке требует некоторого объяснения.
** атрибут ** объекта - это переменная или метод, связанный с ним.
В этом случае значение `t` равно` none`, которое представляет собой объект `nonetype`, который не имеет атрибута с именем` remove`, поэтому результатом является `attributeerror.

Если вы видите подобное сообщение об ошибке, вы должны оглянуться назад через программу и посмотреть, не неправильно ли вы вызвали метод списка.
