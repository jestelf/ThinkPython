# Заключительные мысли


Учиться программировать непросто, но если вы дошли до этого места, вы уже неплохо начали.
Теперь у меня есть несколько предложений, как вы можете продолжать учиться и применять полученные знания.

Эта книга предназначена как общее введение в программирование, поэтому мы не сосредотачивались на конкретных приложениях.
В зависимости от ваших интересов есть множество областей, где вы можете применить свои новые навыки.

Если вам интересна Data Science, вот три моих книги, которые вам могут понравиться:

* *Think Stats: Exploratory Data Analysis*, O'Reilly Media, 2014.

* *Think Bayes: Bayesian Statistics in Python*, O'Reilly Media, 2021.

* *Think DSP: Digital Signal Processing in Python*, O'Reilly Media, 2016.


Если вас интересует физическое моделирование и сложные системы, вам могут понравиться:

* *Modeling and Simulation in Python: An Introduction for Scientists and Engineers*, No Starch Press, 2023.

* *Think Complexity: Complexity Science and Computational Modeling*, O'Reilly Media, 2018.

В этих книгах используются библиотеки NumPy, SciPy, pandas и другие инструменты Python для науки о данных и научных вычислений.


Эта книга пытается найти баланс между общими принципами программирования и деталями Python.
В результате она не включает каждую особенность языка.
Чтобы узнать больше о Python и получить хорошие советы по его использованию, рекомендую *Fluent Python: Clear, Concise, and Effective Programming*, второе издание, Luciano Ramalho, O'Reilly Media, 2022.

После введения в программирование обычный следующий шаг — изучение структур данных и алгоритмов.
У меня есть рабочий черновик на эту тему под названием *Data Structures and Information Retrieval in Python*.
Бесплатную электронную версию можно получить на Green Tea Press по адресу <https://greenteapress.com>.


Работая над более сложными программами, вы столкнётесь с новыми трудностями.
Возможно, будет полезно освежить в памяти разделы этой книги о поиске и исправлении ошибок.
В частности, вспомните шесть 'R' отладки из [главы 12](section_debugging_12): чтение, запуск, размышление, резиновая уточка, отступление и отдых.

В этой книге предлагаются инструменты для отладки, включая функции `print` и `repr`, функцию `structshape` в [главе 11](section_debugging_11) и встроенные функции `isinstance`, `hasattr` и `vars` из [главы 14](section_debugging_14).


Книга также предлагает инструменты для тестирования программ, включая оператор `assert`, модуль `doctest` и модуль `unittest`.
Включение тестов в программы — один из лучших способов предотвращать и обнаруживать ошибки, экономя время на отладке.

Но лучший вид отладки — тот, который вам не приходится делать.
Если вы будете использовать инкрементальный подход к разработке, описанный в [главе 6](section_incremental), и тестировать код по ходу дела, то допустите меньше ошибок и быстрее найдёте те, что всё же появятся.
Также помните об инкапсуляции и обобщении из [главы 4](section_encapsulation), что особенно полезно при разработке кода в Jupyter-ноутбуках.


На протяжении всей книги я предлагал способы использовать виртуальных ассистентов, чтобы учиться, программировать и отлаживать код.
Надеюсь, эти инструменты оказываются вам полезными.

В дополнение к виртуальным ассистентам вроде ChatGPT вы можете попробовать инструмент типа Copilot, который дописывает код по мере набора.
Изначально я не рекомендовал эти инструменты, потому что они могут быть слишком сложными для начинающих, но сейчас вы можете их попробовать.

Эффективное использование ИИ-тулов требует экспериментирования и размышлений, чтобы найти рабочий процесс, который подойдёт именно вам.
Если вам кажется неудобным копировать код из ChatGPT в Jupyter, возможно, вам больше понравится Copilot.
Но умственная работа по составлению запросов и интерпретации ответов может быть так же ценна, как и сгенерированный код — по аналогии с отладкой при помощи резиновой уточки.


Набираясь опыта, вы можете захотеть освоить другие среды разработки.
Я считаю, что Jupyter — хорошее место для старта, но это сравнительно новая среда и она не столь широко распространена, как традиционные интегрированные среды разработки (IDE).
Для Python наиболее популярны PyCharm и Spyder — а также Thonny, который часто рекомендуют новичкам.
Другие IDE, такие как Visual Studio Code и Eclipse, работают и с другими языками программирования.
Более простой вариант — писать программы на Python в любом текстовом редакторе.

Продолжая свой путь в программировании, вам не обязательно идти в одиночку!
Если вы живёте в городе или рядом, вероятно, есть местная группа пользователей Python, к которой можно присоединиться.
Эти группы обычно дружелюбны к новичкам, так что не бойтесь.
Если поблизости нет такой группы, возможно, вы сможете участвовать в мероприятиях удалённо.
Также следите за региональными конференциями по Python.


Один из лучших способов улучшить свои навыки программирования — выучить другой язык.
Если вас интересует статистика и Data Science, можно изучить R.
Но особенно рекомендую функциональный язык вроде Racket или Elixir.
Функциональное программирование требует иного подхода к мышлению, что меняет ваше представление о программах.

Удачи!
