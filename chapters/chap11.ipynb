## КУТУ - это как списки

Крупель - это последовательность значений. Значения могут быть любым типом, и они индексируются целыми числами, поэтому кортежи очень похожи на списки.
Важным отличием является то, что кортежи неизменны.

Чтобы создать кортеж, вы можете написать список значений, разделяемый запятыми.

```python
t = 'l', 'u', 'p', 'i', 'n'
type(t)
```

Хотя в этом нет необходимости, это обычно прикладывает в скобки в скобках.

```python
t = ('l', 'u', 'p', 'i', 'n')
type(t)
```

Чтобы создать кортеж с одним элементом, вы должны включить окончательную запятую.

```python
t1 = 'p',
type(t1)
```

Единственное значение в скобках - это не кортеж.

```python
t2 = ('p')
type(t2)
```

Другим способом создания кортежа является встроенная функция `tuple '. Без
Аргумент, это создает пустой кортеж.

```python
t = tuple()
t
```

Если аргумент представляет собой последовательность (строка, список или кортеж), результатом является
Круп с элементами последовательности.

```python
t = tuple('lupin')
t
```

Поскольку `tuple`-это имя встроенной функции, вы должны избегать использования ее в качестве имени переменной.

Большинство операторов списка также работают с кортежами.
Например, оператор кронштейна индексирует элемент.

```python
t[0]
```

И оператор среза выбирает ряд элементов.

```python
t[1:3]
```

Оператор `+` объединяет платы.

```python
tuple('lup') + ('i', 'n')
```

И оператор `*` дублирует кортеж на определенное количество раз.

```python
tuple('spam') * 2 
```

Функция «отсортированная» работает с кортежами, но результатом является список, а не кортеж.

```python
sorted(t)
```

Функция `reversed` также работает с кортежами.

```python
reversed(t)
```

Результатом является объект «обратный», который мы можем преобразовать в список или кортеж.

```python
tuple(reversed(t))
```

Основываясь на примерах до сих пор, может показаться, что кортежи такие же, как и списки.

## назначение кортежей

Вы можете поставить кортеж с переменными на левую сторону назначения и кортеж значений справа.

```python
a, b = 1, 2
```

Значения присваиваются переменным слева направо - в этом примере `a` получает значение` 1`, а `B` получает значение` 2`.
Мы можем отобразить результаты подобными:

```python
a, b
```

В более общем плане, если левая сторона назначения является кортежом, правая сторона может быть какой -либо последовательности - строка, список или кортеж. 
Например, чтобы разделить адрес электронной почты на имя пользователя и домен, вы можете написать:

```python
email = 'monty@python.org'
username, domain = email.split('@')
```

Возвращаемое значение из `split` - это список с двумя элементами - первый элемент назначается« имя пользователя », второе - домен '.

```python
username, domain
```

Количество переменных слева и количество значений на
Правильно должно быть одинаковым - иначе вы получаете `valueError '.

```python
%%expect ValueError
a, b = 1, 2, 3
```

Назначение кортежей полезно, если вы хотите поменять значения двух переменных.
С обычными заданиями вы должны использовать временную переменную, например:

```python
temp = a
a = b
b = temp
```

Это работает, но с назначением кортежей мы можем сделать то же самое без временной переменной.

```python
a, b = b, a
```

Это работает, потому что все выражения на правой стороне оцениваются перед какими -либо заданиями.

Мы также можем использовать назначение кортежа в операторе `for`.
Например, чтобы пройти через элементы в словаре, мы можем использовать метод «элементов».

```python
d = {'one': 1, 'two': 2}

for item in d.items():
    key, value = item
    print(key, '->', value)
```

Каждый раз через цикл «Item» назначается кортеж, который содержит ключ и соответствующее значение.

Мы можем написать эту петлю более кратко, как это:

```python
for key, value in d.items():
    print(key, '->', value)
```

Каждый раз через цикл, ключ и соответствующее значение назначаются непосредственно на `key` и` value`.

## Аргумент упаковка

Функции могут принимать переменное количество аргументов. 
Имя параметра, которое начинается с оператора `*` упаковывает ** аргументы в кортеж.
Например, следующая функция принимает любое количество аргументов и вычисляет их среднее арифметику, то есть их сумму, разделенную на количество аргументов.

```python
def mean(*args):
    return sum(args) / len(args)
```

Параметр может иметь любое имя, которое вам нравится, но `args 'является обычным.
Мы можем назвать функцию так:

```python
mean(1, 2, 3)
```

Если у вас есть последовательность значений, и вы хотите передать их в функцию в качестве нескольких аргументов, вы можете использовать оператор `*` для ** распаковывать ** кортеж.
Например, «Divmod» принимает ровно два аргумента - если вы передаете кортеж в качестве параметра, вы получаете ошибку.

```python
%%expect TypeError
t = (7, 3)
divmod(t)
```

Несмотря на то, что кортеж содержит два элемента, он считается единственным аргументом.
Но если вы распаковываете кортеж, его рассматривают как два аргумента.

```python
divmod(*t)
```

Упаковка и распаковка могут быть полезны, если вы хотите адаптировать поведение существующей функции.
Например, эта функция принимает любое количество аргументов, удаляет самое низкое и самое высокое и вычисляет среднее значение для остальных.

```python
def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)
```

Сначала он использует `min_max`, чтобы найти самые низкие и самые высокие элементы.
Затем он преобразует «args» в список, чтобы использовать метод `remove`.
Наконец, он распаковывает список, поэтому элементы передаются в «среднее» в виде отдельных аргументов, а не как один список.

Вот пример, который показывает эффект.

```python
mean(1, 2, 3, 10)
```

```python
trimmed_mean(1, 2, 3, 10)
```

Этот вид «подстриженного» среднего значения используется в некоторых видах спорта с субъективным судейством - например, дайвинг и гимнастика - чтобы уменьшить эффект судьи, чей счет отклоняется от других.

## Zip

Клетки полезны для зацикливания через элементы двух последовательностей и выполнения операций на соответствующих элементах.
Например, предположим, что две команды играют в серию из семи игр, и мы записываем их результаты в двух списках, по одной для каждой команды.

```python
scores1 = [1, 2, 4, 5, 1, 5, 2]
scores2 = [5, 5, 2, 2, 5, 2, 3]
```

Давайте посмотрим, сколько игр выиграла каждая команда.
Мы используем `Zip`, который представляет собой встроенную функцию, которая принимает две или более последовательности и возвращает ** Zip-объект **, так называемый, потому что он соединяет элементы последовательностей, таких как зубы молнии.

```python
zip(scores1, scores2)
```

Мы можем использовать объект ZIP, чтобы пройти через значения в последовательностях пары.

```python
for pair in zip(scores1, scores2):
     print(pair)
```

Каждый раз через петлю, «Пару» назначают кортеж с результатами.
Таким образом, мы можем назначить оценки переменным и подсчитать победы для первой команды, например, это:

```python
wins = 0
for team1, team2 in zip(scores1, scores2):
    if team1 > team2:
        wins += 1
        
wins
```

К сожалению, первая команда выиграла только три игры и проиграла сериал.

Если у вас есть два списка и вам нужен список пар, вы можете использовать `Zip` и` list`.

```python
t = list(zip(scores1, scores2))
t
```

Результатом является список кортежей, поэтому мы можем получить результат последней игры, как это:

```python
t[-1]
```

Если у вас есть список ключей и список значений, вы можете использовать `Zip` и` dict ', чтобы сделать словарь.
Например, вот как мы можем сделать словарь, который отображает из каждой буквы в ее положение в алфавите.

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
numbers = range(len(letters))
letter_map = dict(zip(letters, numbers))
```

Теперь мы можем посмотреть букву и получить его индекс в алфавите.

```python
letter_map['a'], letter_map['z']
```

В этом отображении индекс «a» - «0`», а индекс «Z'» - `25`.

Если вам нужно пройти через элементы последовательности и их индексов, вы можете использовать встроенную функцию `перечислить '.

```python
enumerate('abc')
```

Результатом является ** перечисливший объект **, который проходит через последовательность пар, где каждая пара содержит индекс (начиная с 0) и элемент из данной последовательности.

```python
for index, element in enumerate('abc'):
    print(index, element)
```

## Сравнение и сортировка

Реляционные операторы работают с кортежами и другими последовательностями.
Например, если вы используете оператор «<` `с кортежами, он начинается с сравнения первого элемента из каждой последовательности».
Если они равны, это переходит к следующей паре элементов, и так далее, пока не найдет пару, которая отличается.

```python
(0, 1, 2) < (0, 3, 4)
```

Последующие элементы не рассматриваются - даже если они действительно большие.

```python
(0, 1, 2000000) < (0, 3, 4)
```

Этот способ сравнения кортежей полезен для сортировки списка кортежей или поиска минимального или максимума.
В качестве примера, давайте найдем наиболее распространенное письмо одним словом.
В предыдущей главе мы написали `value_counts`, который берет строку и возвращает словарь, который отображает из каждой буквы в количество раз.

```python
def value_counts(string):
    counter = {}
    for letter in string:
        if letter not in counter:
            counter[letter] = 1
        else:
            counter[letter] += 1
    return counter
```

Вот результат для строки `'banana''.

```python
counter = value_counts('banana')
counter
```

Только с тремя элементами мы можем легко увидеть, что самая частая буква - «A», которая появляется три раза.
Но если бы было больше предметов, было бы полезно разортировать их автоматически.

Мы можем получить элементы от «стойки», как это.

```python
items = counter.items()
items
```

Результатом является объект `dict_items`, который ведет себя как список кортежей, поэтому мы можем сортировать его так.

```python
sorted(items)
```

Поведение по умолчанию состоит в том, чтобы использовать первый элемент из каждого корзина для сортировки списка и использовать второй элемент для разрыва связей.

Однако, чтобы найти элементы с наибольшим количеством количества, мы хотим использовать второй элемент для сортировки списка.
Мы можем сделать это, написав функцию, которая берет кортеж и возвращает второй элемент.

```python
def second_element(t):
    return t[1]
```

Затем мы можем передать эту функцию в «сортированное» в качестве необязательного аргумента, называемого «Key», который указывает на то, что эта функция должна использоваться для вычисления клавиши ** Sort ** для каждого элемента.

```python
sorted_items = sorted(items, key=second_element)
sorted_items
```

Ключ сорта определяет порядок элементов в списке.
Сначала появляется письмо с самым низким подсчетом, и буква с наивысшим количеством подсчета появляется последней.
Таким образом, мы можем найти самое распространенное письмо, как это.

```python
sorted_items[-1]
```

Если мы хотим только максимум, нам не нужно сортировать список.
Мы можем использовать `max`, который также принимает` key 'в качестве необязательного аргумента.

```python
max(items, key=second_element)
```

Чтобы найти букву с самым низким подсчетом, мы могли бы использовать «min» таким же образом.

## инвертирование словаря

Предположим, вы хотите инвертировать словарь, чтобы вы могли найти значение и получить соответствующий ключ.
Например, если у вас есть счетчик слов, который отображает от каждого слова до того, как оно появляется, вы можете сделать словарь, который отображает из целых чисел в слова, которые появляются в течение этого количества раз.

Но есть проблема - ключи в словаре должны быть уникальными, но значения нет. Например, в счетчике слова может быть много слов с тем же количеством.

Таким образом, одним из способов инвертировать словарь является создание нового словаря, в котором значения являются списками клавиш из оригинала.
В качестве примера, давайте подсчитаем буквы в «попугаи».

```python
d =  value_counts('parrot')
d
```

Если мы инвертируем этот словарь, результат должен быть `{1: ['p', 'a', 'o', 't'], 2: ['r']}`, что указывает на то, что буквы, которые когда -то появляются, - «p'`», `a'``,` 'O'' и `'t', и буква, чем появляется, - это два раза - это` r '.

Следующая функция принимает словарь и возвращает его обратный в качестве нового словаря.

```python
def invert_dict(d):
    new = {}
    for key, value in d.items():
        if value not in new:
            new[value] = [key]
        else:
            new[value].append(key)
    return new
```

Оператор «для» проходит через ключи и значения в `d`.
Если значение еще не в новом словаре, оно добавлено и связано с списком, который содержит один элемент.
В противном случае он добавляется в существующий список.

Мы можем проверить это так:

```python
invert_dict(d)
```

И мы получаем результат, который мы ожидали.

Это первый пример, где мы видели, где значения в словаре являются списками.

## отладка

Списки, словари и кортежи - это ** структуры данных **.
В этой главе мы начинаем видеть составные структуры данных, такие как списки кортежей, или словари, которые содержат кортежи в качестве ключей и списков в качестве значений.
Составные структуры данных полезны, но они подвержены ошибкам, вызванным, когда структура данных имеет неправильный тип, размер или структуру.
