## Аргументы

При вызове функции выражение в скобках называется * * аргументом**.
Обычно я объясняю, почему, но в этом случае техническое значение термина почти не имеет ничего общего с общим значением этого слова, поэтому я даже не буду пытаться.

Некоторые из функций, которые мы видели до сих пор, принимают только один аргумент, например, «int».

Некоторые берут два, например «math.pow».

Некоторые могут принимать дополнительные аргументы, которые являются необязательными. 
Например, `int` может принимать второй аргумент, указывающий основание числа.

Последовательность цифр «101» в основании 2 представляет собой число 5 в основании 10.

`round` также принимает необязательный второй аргумент, который является числом знаков после запятой, до которого нужно округлить.

Некоторые функции могут принимать любое количество аргументов, например `print`.

Если вы вызываете функцию и предоставляете слишком много аргументов, это `TypeError`.

Если вы предоставляете слишком мало аргументов, это также `TypeError`.

И если вы предоставляете аргумент с типом, который функция не может обработать, это тоже `TypeError`.

Такая проверка может раздражать, когда вы начинаете, но она помогает обнаруживать и исправлять ошибки.

## Отладка

В программе могут возникать три вида ошибок: синтаксические ошибки, ошибки времени выполнения и семантические ошибки. Полезно различать их, чтобы быстрее находить.

* **Синтаксическая ошибка**: "Синтаксис" относится к структуре программы и правилам этой структуры. Если где-то в программе есть синтаксическая ошибка, Python не запускает программу и сразу выводит сообщение об ошибке.

* **Ошибка времени выполнения**: если синтаксических ошибок нет, программа может начать выполняться. Но если что-то идёт не так, Python выводит сообщение об ошибке и останавливается. Такая ошибка называется ошибкой времени выполнения, или **исключением**, потому что она означает возникновение исключительной ситуации.

* **Семантическая ошибка**: третий тип ошибок связан со смыслом. Если в программе есть семантическая ошибка, она выполняется без сообщений об ошибке, но делает не то, что вы ожидали. Выявить такие ошибки сложно, потому что приходится анализировать результаты работы программы и пытаться понять, что она делает.

Как мы видели, недопустимое имя переменной является синтаксической ошибкой.

Если вы используете оператор с типом, который он не поддерживает, это ошибка времени выполнения.

Наконец, вот пример семантической ошибки.

Когда это выражение вычисляется, оно не выдает сообщение об ошибке, поэтому нет синтаксической ошибки или ошибки во время выполнения.
Но результат не является средним из «1» и «3», поэтому программа неверна.
Это семантическая ошибка, потому что программа работает, но не делает то, что задумано.

## Глоссарий

**переменная:**
Имя, которое ссылается на значение.

**оператор присваивания:**
Инструкция, которая присваивает значение переменной.

**диаграмма состояния:**
Графическое представление набора переменных и связанных с ними значений.

**ключевое слово:**
Специальное слово, используемое для задания структуры программы.

**инструкция import:**
Инструкция, которая читает файл модуля, чтобы мы могли использовать содержащиеся в нём переменные и функции.

**модуль:**
Файл, который содержит код на Python, включая определения функций и иногда другие инструкции.

**оператор точки:**
Оператор `.` , используемый для доступа к функции в другом модуле путём указания имени модуля, за которым следует точка и имя функции.

**вычислить:**
Выполнить операции в выражении, чтобы получить значение.

**инструкция:**
Одна или несколько строк кода, представляющие команду или действие.

**выполнить:**
Запустить инструкцию и сделать то, что она предписывает.

**аргумент:**
Значение, передаваемое функции при её вызове.

**комментарий:**
Текст в программе, который предоставляет информацию, но не влияет на выполнение.

**ошибка времени выполнения:**
Ошибка, из-за которой программа выводит сообщение об ошибке и завершается.

**исключение:**
Ошибка, обнаруженная во время выполнения программы.

**семантическая ошибка:**
Ошибка, из-за которой программа делает не то, что нужно, но сообщение об ошибке не выводится.
