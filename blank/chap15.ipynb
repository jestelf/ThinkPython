# Классы и методы

Python-это ** объектно-ориентированный язык **, то есть он предоставляет функции, которые поддерживают объектно-ориентированное программирование, которое имеет эти определяющие характеристики:

- Большая часть вычислений выражена с точки зрения операций на объектах.

- Объекты часто представляют вещи в реальном мире, и методы часто соответствуют способам взаимодействия вещей в реальном мире.

- Программы включают определения класса и метода.

Например, в предыдущей главе мы определили класс `Time`, который соответствует тому, как люди записывают время суток, и мы определили функции, которые соответствуют тому, что люди делают со временем.
Но не было явного соединения между определением класса `Time` и следующими определениями функций.
Мы можем явно сделать соединение, переписывая функцию как ** метод **, который определяется внутри определения класса.

## метод `__str__`

Когда вы пишете метод, вы можете выбрать практически любое имя, которое вы хотите.
Однако некоторые имена имеют особые значения.
Например, если у объекта есть метод с именем `__str__`, Python использует этот метод для преобразования объекта в строку.
Например, вот метод `__str__` для объекта времени.

Этот метод аналогичен `print_time`, из предыдущей главы, за исключением того, что он возвращает строку, а не печатает ее.

Вы можете вызвать этот метод обычным способом.

Но Python также может вызвать его для вас.
Если вы используете встроенную функцию `str` для преобразования объекта `Time` в строку, Python использует метод `__str__` в классе `Time`.

И это делает то же самое, если вы печатаете объект `Time`.

Методы, такие как `__str__`, называются ** Специальные методы **.
Вы можете идентифицировать их, потому что их имена начинаются и заканчиваются двумя подставками.

## перегрузка оператора

Определяя другие специальные методы, вы можете указать поведение
Операторы на определяемых программистом типах.Например, если вы определите
Метод с именем `__add__` для класса `Time`, вы можете использовать `+`

Вот метод `__add__`.

Мы можем использовать это так.

Когда мы запускаем эти три строки кода, многое происходит:

* Когда мы создаем создание объекта `Time`, используется метод `__init__`.

* Когда мы используем оператор `+` с объектом `Time`, его метод `__add__` вызывается.

* И когда мы печатаем объект `Time`, его метод `__str__` вызывается.

Изменение поведения оператора, так что он работает с типами, определенными программистом, называется ** Оператор перегрузка **.
Для каждого оператора, как `+`, существует соответствующий специальный метод, например, `__add__`.

## отладка

Объект `Time` действителен, если значения `minute` и `second` находятся между `0` и `60` - включая `0`, но не `60` - и если `hour` положительно.
Кроме того, `hour` и `minute` должны быть целочисленными значениями, но мы можем разрешить `second` иметь часть дробной части.
Требования, подобные этим, называются ** инвариантами **, потому что они всегда должны быть правдой.
Чтобы выразить это по -другому, если они неправда, что -то пошло не так.

Написание кода для проверки инвариантов может помочь обнаружить ошибки и найти их причины.
Например, у вас может быть метод, такой как `minute`1, который занимает объект времени и возвращает `False`, если он нарушает инвариант.

Затем, в начале каждого метода, вы можете проверить аргументы, чтобы убедиться, что они действительны.

Заявление `assert` оценивает следующее выражение.Если результат `True`, он ничего не делает;Если результат `False`, он вызывает `AssertionError`.

## Глоссарий

** объектно-ориентированный язык: **
Язык, который предоставляет функции для поддержки объектно-ориентированного программирования, особенно определяемых пользователями типов.

** Метод: **
Функция, которая определяется внутри определения класса и вызывается в случаях этого класса.

** приемник: **
Объект, на который используется метод.

** Статический метод: **
Метод, который можно вызвать без объекта в качестве приемника.

** Метод экземпляра: **
Метод, который должен быть вызван объектом в качестве приемника.

** Специальный метод: **
Метод, который меняет способ работы операторов и некоторых функций с объектом.

** Оператор перегрузка: **
Процесс использования специальных методов для изменения способа операторов с пользовательскими типами.

** инвариант: **
Условие, которое всегда должно быть правдой во время выполнения программы.
