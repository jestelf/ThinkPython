# Предисловие



## Для кого эта книга?

Если вы хотите научиться программировать, вы обратились по адресу.
Python — один из лучших языков для новичков и одновременно один из самых востребованных навыков.

Вы также выбрали подходящее время: сейчас учиться программировать, пожалуй, проще, чем когда-либо.
С виртуальными помощниками вроде ChatGPT вам не обязательно учиться в одиночку.
На протяжении всей книги я буду предлагать способы использовать эти инструменты, чтобы ускорить обучение.

Эта книга предназначена прежде всего для тех, кто никогда не программировал, и для людей с опытом в других языках.
Если у вас уже есть серьёзный опыт работы с Python, первые главы могут показаться слишком медленными.

Одна из трудностей обучения программированию состоит в том, что нужно изучить *два* языка: сам язык программирования и словарь терминов, которыми мы описываем программы.
Если вы выучите только сам язык, вам будет сложно разбираться в сообщениях об ошибках, читать документацию, общаться с другими людьми и пользоваться виртуальными помощниками.
Если вы уже немного программировали, но не освоили эту вторую терминологию, надеюсь, эта книга поможет вам.

## Цели книги

Пишя эту книгу, я старался аккуратно использовать терминологию.
Я даю определение каждому термину, когда он впервые появляется.
В конце каждой главы есть глоссарий, где перечислены введённые термины.

Я также старался быть кратким.
Чем меньше усилий требуется для чтения книги, тем больше сил останется на программирование.

Но научиться программировать только читая книгу невозможно — нужно практиковаться.
Поэтому в конце каждой главы есть упражнения, где вы можете потренироваться.

Если читать внимательно и регулярно выполнять упражнения, вы будете продвигаться вперёд.
Но предупреждаю заранее: учиться программированию непросто, и даже опытные программисты иногда испытывают разочарование.
По ходу дела я буду предлагать стратегии, помогающие писать правильные программы и исправлять ошибки.

## Как работать с книгой

Каждая глава этой книги опирается на предыдущие, поэтому лучше читать их по порядку и не переходить далее, пока не выполните упражнения.

В первых шести главах рассматриваются базовые элементы: арифметика, условные операторы и циклы.
Также в них вводится важнейшая концепция программирования — функции, а также мощный способ их применения — рекурсия.

Седьмая и восьмая главы посвящены строкам — они могут представлять буквы, слова и предложения — и алгоритмам работы с ними.

Глава 9—12 знакомят с основными структурами данных Python — списками, словарями и кортежами, которые позволяют писать эффективные программы.
В 12-й главе описаны алгоритмы анализа текста и генерации нового текста.
Такие алгоритмы лежат в основе больших языковых моделей (LLM), поэтому эта глава даст вам представление о том, как работают инструменты вроде ChatGPT.

Глава 13 рассказывает о способах долговременного хранения данных — файлах и базах данных.
В качестве упражнения вы сможете написать программу, которая ищет в файловой системе дубликаты файлов.

Главы 14–17 вводят объектно-ориентированное программирование (ООП) — способ организовывать программы и данные, с которыми они работают.
Многие библиотеки Python написаны в объектно-ориентированном стиле, поэтому эти главы помогут понять их устройство и создавать собственные объекты.

Цель книги — не покрыть весь язык Python полностью.
Я сосредоточился на той части языка, которая дает наибольшие возможности при минимальном наборе концепций.
Тем не менее в Python есть много возможностей, которые помогут эффективно решать типовые задачи.
Некоторые из них рассматриваются в главе 18.

Наконец, в главе 19 я делюсь завершающими мыслями и советами по продолжению изучения программирования.

## Что нового в третьем издании?

Наибольшие изменения в этом издании связаны с двумя новыми технологиями — Jupyter‑ноутбуками и виртуальными ассистентами.

Каждая глава этой книги представляет собой Jupyter‑ноутбук — документ, содержащий и обычный текст, и код.
Мне это упрощает написание кода, его тестирование и поддержание согласованности с текстом.
Для вас это означает возможность запускать код, изменять его и выполнять упражнения прямо в одном месте.
Инструкции по работе с ноутбуками приведены в первой главе.

Другое важное нововведение — советы по работе с виртуальными помощниками вроде ChatGPT и использованию их для ускорения обучения.
Когда предыдущее издание вышло в 2016 году, предшественники этих инструментов были гораздо менее полезны, и большинство людей о них не знало.
Теперь же они стали стандартным инструментом разработки ПО, и я думаю, что они окажут революционное влияние на обучение программированию и многим другим вещам.

Остальные изменения книги вызваны моими сожалениями по поводу второго издания.

Прежде всего я недостаточно уделил внимания тестированию программного обеспечения.
Это и тогда было досадным упущением, а с появлением виртуальных ассистентов автоматическое тестирование стало ещё важнее.
Поэтому в этом издании представлены наиболее популярные средства тестирования Python — `doctest` и `unittest`, а также несколько упражнений для их практики.

Моё другое сожаление заключалось в том, что упражнения во втором издании были неравноценными: одни интереснее других, а некоторые слишком сложны.
Переход на Jupyter‑ноутбуки помог мне разработать и проверить более увлекательную и эффективную последовательность упражнений.

В этой редакции последовательность тем почти та же, но я переставил несколько глав и объединил две короткие главы в одну.
Кроме того, я расширил раздел о строках, добавив регулярные выражения.

Несколько глав используют черепашью графику.
В предыдущих изданиях я применял модуль `turtle` из Python, но, к сожалению, он не работает в Jupyter‑ноутбуках.
Поэтому я заменил его новым модулем черепахи, пользоваться которым должно быть проще.

Наконец, я переписал значительную часть текста, прояснив одни места и сократив другие, где мог быть лаконичнее.

Я очень горжусь этим новым изданием — надеюсь, оно вам понравится!

## Начало работы

Для большинства языков программирования, включая Python, существует множество инструментов для написания и запуска программ.
Эти инструменты называют интегрированными средами разработки (IDE).
В общем случае IDE бывают двух видов:

* Одни работают с файлами, содержащими код, и предоставляют средства для их редактирования и запуска.

* Другие ориентированы на работу с ноутбуками — документами, в которых есть текст и код.

Новичкам я рекомендую начать с среды для работы с ноутбуками, такой как Jupyter.

Ноутбуки из этой книги доступны в онлайн‑репозитории <https://allendowney.github.io/ThinkPython>.

Есть два способа их использовать:

* Вы можете скачать ноутбуки и запускать их на своём компьютере. Для этого нужно установить Python и Jupyter — это несложно, но если вы только начинаете изучать Python, может расстраивать трата времени на установку ПО.

* Другой вариант — работать с ноутбуками в Colab. Это среда Jupyter, которая запускается в браузере, поэтому ничего не нужно устанавливать. Colab управляется Google и бесплатен.

Если вы только начинаете, настоятельно советую начинать с Colab.
