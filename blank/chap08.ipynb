## Строковые методы

Строки предоставляют методы, которые выполняют различные полезные операции.
Метод похож на функцию - он принимает аргументы и возвращает значение - но синтаксис отличается.
Например, метод `upper` принимает строку и возвращает новую строку со всеми заглавными буквами.

Вместо синтаксиса функции `upper(word)` он использует синтаксис метода `word.upper()`.

Это использование оператора DOT указывает имя метода `upper`, и имя строки для применения метода к `word`.
Пустые скобки указывают, что этот метод не принимает аргументов.

Вызов метода называется ** вызов **;В этом случае мы бы сказали, что мы вызываем `upper` на `word`.

## отладка

Когда вы читаете и пишете файлы, отладка может быть сложной.
Если вы работаете в ноутбуке Jupyter, вы можете использовать ** команды оболочки **, чтобы помочь.
Например, чтобы отобразить первые несколько строк файла, вы можете использовать команду `!head`, например:

Начальная восклицательная точка, `!`, указывает на то, что это команда оболочки, которая не является частью Python.
Чтобы отобразить последние несколько строк, вы можете использовать `!tail`.

Когда вы работаете с большими файлами, отладка может быть трудной, потому что может быть слишком много выводов, чтобы проверить вручную.
Хорошая стратегия отладки состоит в том, чтобы начать только с части файла, заработать программу, а затем запустить ее со всем файлом.

Чтобы сделать небольшой файл, который содержит часть более крупного файла, мы можем снова использовать `!head` с оператором перенаправления, `>`, что указывает на то, что результаты должны быть записаны в файл, а не отображаются.

По умолчанию `!head` считывает первые 10 строк, но ему требуется дополнительный аргумент, который указывает количество строк для чтения.

## Глоссарий

**последовательность:**
Упорядоченный сбор значений, где каждое значение определяется целочисленным индексом.

**характер:**
Элемент строки, включая буквы, числа и символы.

** Индекс: **
Целое значение, используемое для выбора элемента в последовательности, такой как символ в строке.В индексах Python начинаются с `0`.

** Slice: **
Часть строки, указанная в диапазоне индексов.

** пустая строка: **
Строка, которая не содержит символов и имеет длину `0`.

**объект:**
То, на что может ссылаться переменная.Объект имеет тип и значение.

** Неизвываемо: **
Если элементы объекта не могут быть изменены, объект неизменен.

** Призыв: **
Выражение - или часть выражения - которое вызывает метод.

** Регулярное выражение: **
Последовательность символов, которые определяют шаблон поиска.

**шаблон:**
Правило, которое указывает требования, которые должна соответствовать строке, чтобы составить матч.

** Строка заменить: **
Замена строки или часть строки, с другой строкой.

** Команда оболочки: **
Заявление на языке оболочки, которое является языком, используемым для взаимодействия с операционной системой.
