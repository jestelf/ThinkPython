## Повторение

Если мы хотим вывести больше одного куплета, можно использовать оператор `for`.
Вот простой пример.

Первая строка — заголовок, который заканчивается двоеточием.
Вторая строка — тело, которое должно иметь отступ.

Заголовок начинается с ключевого слова `for`, новой переменной `i` и ещё одного ключевого слова `in`. 
Он использует функцию `range`, чтобы создать последовательность из двух значений: `0` и `1`.
В Python обычно начинают счёт с `0`.

Когда выполняется оператор `for`, он присваивает переменной `i` первое значение из `range`, а затем выполняет `print` в теле, выводя `0`.

Дойдя до конца тела, выполнение возвращается к заголовку, поэтому эта конструкция называется **циклом**.
Во второй проход цикла в `i` помещается следующее значение из `range`, которое выводится.
Затем, поскольку это последнее значение из `range`, цикл завершается.

Вот как можно с помощью цикла `for` напечатать два куплета песни.

Внутри функции можно поместить цикл `for`.
Например, `print_n_verses` принимает параметр `n`, который должен быть целым числом, и выводит указанное количество куплетов. 

В этом примере переменная `i` в теле цикла не используется, но имя переменной в заголовке всё равно должно быть.

## Диаграммы стека

Чтобы отслеживать, какие переменные где доступны, иногда полезно рисовать **диаграмму стека**. 
Как и диаграммы состояний, диаграммы стека показывают значения каждой переменной, а также функцию, к которой она относится.

Каждая функция представлена **кадром**.
Кадр — это прямоугольник с именем функции снаружи и параметрами и локальными переменными внутри.

Вот диаграмма стека для предыдущего примера.

````python
from diagram import make_frame, Stack

d1 = dict(line1=line1, line2=line2)
frame1 = make_frame(d1, name='__main__', dy=-0.3, loc='left')

d2 = dict(part1=line1, part2=line2, cat=line1+line2)
frame2 = make_frame(d2, name='cat_twice', dy=-0.3, 
                    offsetx=0.03, loc='left')

d3 = dict(string=line1+line2)
frame3 = make_frame(d3, name='print_twice', 
                    offsetx=0.04, offsety=-0.3, loc='left')

d4 = {"?": line1+line2}
frame4 = make_frame(d4, name='print', 
                    offsetx=-0.22, offsety=0, loc='left')

stack = Stack([frame1, frame2, frame3, frame4], dy=-0.8)
````

````python
from diagram import diagram, adjust


width, height, x, y = [3.77, 2.9, 1.1, 2.65]
ax = diagram(width, height)
bbox = stack.draw(ax, x, y)
# adjust(x, y, bbox)
````

Кадры расположены в стеке, показывая, какая функция вызвала
какую, и так далее. Снизу вверх: `print` вызвана `print_twice`, её вызвала `cat_twice`, а её — `__main__`, что является особым именем верхнего кадра.
Когда вы создаёте переменную вне какой-либо функции, она принадлежит `__main__`.

## Обратные трассировки

Когда во время выполнения в функции возникает ошибка, Python показывает имя выполнявшейся функции, имя вызвавшей её функции и так далее вверх по стеку.
Чтобы увидеть пример, я определю версию `print_twice`, содержащую ошибку — она пытается вывести `cat`, которая является локальной переменной в другой функции.

А вот что происходит, когда мы запускаем `cat_twice`.

````python
# This cell tells Jupyter to provide detailed debugging information
# when a runtime error occurs, including a traceback.

%xmode Verbose
````

Сообщение об ошибке содержит **обратную трассировку**, в которой показана функция, во время работы которой произошла ошибка, функция, вызвавшая её, и так далее.
В этом примере видно, что `cat_twice` вызвала `print_twice`, и ошибка произошла в `print_twice`.

Порядок функций в трассировке соответствует порядку кадров на диаграмме стека.
Функция, выполнявшаяся в момент ошибки, находится внизу.

## Зачем нужны функции?

Пока может быть непонятно, зачем делить программу на функции.
Есть несколько причин:

-   Создавая новую функцию, вы можете дать название группе инструкций, что делает программу понятнее и упрощает отладку.

-   Функции позволяют уменьшить программу, убрав повторяющийся код. Внесённые изменения нужно делать только в одном месте.

-   Разбивая большую программу на функции, можно отлаживать части по отдельности, а затем собрать их в работающую систему.

-   Хорошо спроектированные функции часто пригодятся в других программах. Написав и отладив их однажды, можно использовать снова.

## Отладка

Отладка может быть раздражающей, но в то же время она сложна, интересна и порой даже увлекательна.
И это один из важнейших навыков, которым стоит овладеть.

В некотором смысле отладка похожа на детективное расследование.
У вас есть улики, и вам нужно восстановить события, которые привели к наблюдаемому результату.

Отладка также похожа на экспериментальную науку.
Получив предположение о причине сбоя, вы изменяете программу и пробуете снова.
Если гипотеза верна, вы можете предсказать результат изменения и приблизиться к рабочей программе.
Если же гипотеза ошибочна, придётся придумать новую.

Для некоторых людей программирование и отладка — это одно и то же: программирование — это процесс постепенной отладки программы, пока она не станет делать то, что нужно.
Идея в том, чтобы начинать с работающей программы и вносить небольшие изменения, отлаживая их по мере продвижения.

Если вы тратите много времени на отладку, это часто означает, что вы пишете слишком много кода до начала тестирования.
Делая более короткие шаги, можно обнаружить, что работа идёт быстрее.

## Глоссарий

**определение функции:**
Инструкция, создающая функцию.

**заголовок:**
 Первая строка определения функции.

**тело:**
 Набор инструкций внутри определения функции.

**объект функции:**
Значение, создаваемое определением функции.
Имя функции — это переменная, которая ссылается на объект функции.

**параметр:**
 Имя, используемое внутри функции для обращения к переданному аргументу.

**цикл:**
 Инструкция, которая выполняет одну или несколько инструкций, часто повторяя их.

**локальная переменная:**
Переменная, объявленная внутри функции и доступная только в этой функции.

**диаграмма стека:**
Графическое представление стека функций, их переменных и значений, на которые они ссылаются.

**кадр:**
 Прямоугольник на диаграмме стека, представляющий вызов функции.
 Он содержит локальные переменные и параметры функции.

**обратная трассировка:**
 Список выполнявшихся функций, выводимый при возникновении исключения.
