# Заключительные мысли

Учиться программировать непросто, но если вы дошли до этого места, у вас уже хороший старт.
Теперь несколько советов, как продолжать учиться и применять полученные знания.

Эта книга является общим введением в программирование, поэтому мы не сосредоточились на конкретных приложениях.
В зависимости от ваших интересов существует множество областей, где вы можете применить новые навыки.

Если вас интересует Data Science, вам могут понравиться три мои книги:

* *Think Stats: Exploratory Data Analysis*, O'Reilly Media, 2014.

* *Think Bayes: Bayesian Statistics in Python*, O'Reilly Media, 2021.

* *Think DSP: Digital Signal Processing in Python*, O'Reilly Media, 2016.

Если вас интересует физическое моделирование и сложные системы, вам могут понравиться следующие книги:

* *Modeling and Simulation in Python: An Introduction for Scientists and Engineers*, No Starch Press, 2023.

* *Think Complexity: Complexity Science and Computational Modeling*, O'Reilly Media, 2018.

В них используются NumPy, SciPy, pandas и другие библиотеки Python для науки о данных и научных вычислений.

Эта книга пытается найти баланс между общими принципами программирования и деталями Python.
В результате она не включает каждую возможность языка Python.
Чтобы узнать больше о Python и получить рекомендации по его применению, я советую книгу *Fluent Python: Clear, Concise, and Effective Programming*, второе издание, Лучиано Рамальо, O'Reilly Media, 2022.

После введения в программирование обычным следующим шагом является изучение структур данных и алгоритмов.
У меня есть работа в процессе на эту тему под названием *Data Structures and Information Retrieval in Python*.
Бесплатная электронная версия доступна в Green Tea Press по адресу <https://greenteapress.com>.

По мере того как вы будете писать более сложные программы, вы столкнетесь с новыми вызовами.
Возможно, вам будет полезно перечитать разделы этой книги о отладке.
В частности, помните о шести R отладки из [главы 12](section_debugging_12): чтение, запуск, размышление, 'резиновая уточка', отход и отдых.

В этой книге также предлагаются инструменты для отладки, включая функции `print` и `repr`, функцию `structshape` из [главы 11](section_debugging_11), а также встроенные функции `isinstance`, `hasattr` и `vars` из [главы 14](section_debugging_14).

Также предлагаются инструменты для тестирования программ: оператор `assert`, модуль `doctest` и модуль `unittest`.
Включение тестов в программу — один из лучших способов предотвратить и обнаружить ошибки, сэкономив время на отладке.

Но лучший вид отладки — это тот, который вам не приходится делать.
Если вы используете процесс инкрементальной разработки, описанный в [главе 6](section_incremental), и тестируете по ходу дела, ошибок будет меньше и вы будете быстрее их находить.
Также помните об инкапсуляции и обобщении из [главы 4](section_encapsulation), что особенно полезно при разработке кода в Jupyter-ноутбуках.

На протяжении этой книги я предлагал способы использовать виртуальных ассистентов для обучения, программирования и отладки.
Надеюсь, эти инструменты оказались полезными.

Помимо ассистентов вроде ChatGPT, вы также можете попробовать такие средства, как Copilot, которое дополняет код по мере набора.
Изначально я не рекомендовал использовать их, так как они могут перегружать новичков, но сейчас вы можете их изучить.

Эффективное применение ИИ-инструментов требует экспериментов и размышлений, чтобы найти подходящий вам рабочий процесс.
Если вам неудобно копировать код из ChatGPT в Jupyter, возможно, вам подойдет Copilot.
Но умственная работа, которую вы проделываете, формулируя запрос и интерпретируя ответ, может быть не менее ценной, чем сам код, который генерирует инструмент, в духе резиновой уточки.

Набираясь опыта программирования, вы можете захотеть опробовать другие среды разработки.
Я считаю Jupyter-ноутбуки хорошим местом для начала, но они относительно новы и не так широко распространены, как традиционные интегрированные среды разработки (IDE).
Для Python наиболее популярны PyCharm и Spyder, а также Thonny, который часто рекомендуют новичкам.
Другие IDE, такие как Visual Studio Code и Eclipse, поддерживают и другие языки программирования.
Или, в качестве более простого варианта, вы можете писать программы на Python в любом текстовом редакторе.

Продолжая путь программиста, не оставайтесь одни!
Если вы живёте в городе или рядом, скорее всего там есть группа пользователей Python, к которой можно присоединиться.
Эти сообщества обычно дружелюбны к новичкам, поэтому не бойтесь.
Если рядом нет такой группы, возможно, вы сможете участвовать в событиях удаленно.
Также обращайте внимание на региональные конференции по Python.

Один из лучших способов улучшить навыки программирования — выучить другой язык.
Если вас интересует статистика и Data Science, возможно, вам стоит изучить R.
Но особенно рекомендую изучить функциональный язык, например Racket или Elixir.
Функциональное программирование требует иного образа мышления, который меняет восприятие программ.

Удачи!
