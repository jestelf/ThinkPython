# Итерации и поиск

В 1939 году Эрнест Винсент Райт опубликовал роман из 50 000 слов под названием *Gadsby*, в котором нет буквы "e". Поскольку "e" — самая распространённая буква в английском языке, написать даже несколько слов, не используя её, сложно.

Мы воспользуемся оператором `in`, чтобы проверить, встречается ли буква в слове, а также познакомимся с программным шаблоном под названием "линейный поиск".

## Обновление переменных

Как вы, возможно, уже выяснили, допустимо присваивать значение одной и той же переменной несколько раз.
Новое присваивание заставляет существующую переменную ссылаться на новое значение (и перестать ссылаться на старое).

Например, вот первоначальное присваивание, которое создаёт переменную.

А вот присваивание, которое изменяет значение переменной.

Следующий рисунок показывает, как эти присваивания выглядят на диаграмме состояний.

Пунктирная стрелка показывает, что `x` больше не ссылается на `5`.
Сплошная стрелка показывает, что теперь она ссылается на `7`.

Распространённый вид присваивания — **обновление**, когда новое значение переменной зависит от старого.

Эта инструкция означает: «возьми текущее значение `x`, прибавь единицу и запиши результат обратно в `x`».

Если попытаться обновить переменную, которой не существует, возникнет ошибка, потому что Python вычисляет выражение справа прежде, чем присвоить значение переменной слева.

Увеличение значения переменной называется **инкрементом**, уменьшение — **декрементом**.
Поскольку эти операции очень распространены, Python предоставляет **операторы составного присваивания**, позволяющие обновлять переменную короче.
Например, оператор `+=` увеличивает переменную на указанную величину.

Для других арифметических операторов тоже есть составные операторы присваивания, включая `-=` и `*=`.

`uses_any` переводит `word` и `letters` в нижний регистр, поэтому работает при любом сочетании заглавных и строчных букв.

Структура `uses_any` похожа на `has_e`.
Она перебирает буквы слова и проверяет их по одной.
Если встречается буква из `letters`, функция немедленно возвращает `True`.
Если цикл завершается, не найдя ни одной, возвращается `False`.

Такой шаблон называется **линейным поиском**.

## Doctest

В [главе 4](section_docstring) мы использовали docstring для документирования функции — то есть чтобы объяснить, что она делает.
Docstring можно также использовать для *тестирования* функции.
Вот версия `uses_any` с docstring, который содержит тесты.

Каждый тест начинается с `>>>`, что используется в некоторых средах Python как приглашение, куда пользователь может вводить код.
В doctest за приглашением следует выражение, обычно вызов функции.
Следующая строка указывает значение, которое должно получиться, если функция работает правильно.

В первом примере `'banana'` содержит букву `'a'`, поэтому результат должен быть `True`.
Во втором примере `'apple'` не содержит ни одной из букв `'xyz'`, поэтому результат должен быть `False`.

## Глоссарий

**переменная цикла:**
Переменная, определённая в заголовке цикла `for`.

**файловый объект:**
Объект, представляющий открытый файл и отслеживающий, какие части файла уже прочитаны или записаны.

**метод:**
 Функция, связанная с объектом и вызываемая через точку.

**обновление:**
Оператор присваивания, который задаёт новое значение существующей переменной, а не создаёт новую.

**инициализация:**
Создание новой переменной и присвоение ей значения.

**инкремент:**
Увеличение значения переменной.

**декремент:**
Уменьшение значения переменной.

**счётчик:**
 Переменная, используемая для подсчёта чего-либо, обычно инициализируется нулём и затем увеличивается.

**линейный поиск:**
Вычислительный шаблон, который проходит по последовательности элементов и останавливается, когда находит искомое.

**успех (pass):**
Тест проходит, если результат соответствует ожидаемому.

**неудача (fail):**
Тест не проходит, если результат отличается от ожидаемого.
