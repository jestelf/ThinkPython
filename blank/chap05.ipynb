## Глоссарий

** Рекурсия: **
Процесс вызова функции, которая в настоящее время выполняет.

** Оператор модуля: **
Оператор, «%», который работает на целых числах и возвращает остаток, когда одно число разделено на другое.

** Логическое выражение: **
Выражение, значение которого либо `true`, либо` false '.

** Реляционный оператор: **
Один из операторов, который сравнивает его операнды: `==`, `! =`, `>`, `<`, `> =` И `<=`.

** Логический оператор: **
Один из операторов, который сочетает в себе логические выражения, в том числе «и», «или» и «Не».

** условное утверждение: **
Оператор, который контролирует поток выполнения в зависимости от какого -либо условия.

**состояние:**
Логическое выражение в условном утверждении, которое определяет, какая ветвь работает.

**блокировать:**
Одно или несколько утверждений отступают, чтобы указать, что они являются частью другого утверждения.

**ветвь:**
Одна из альтернативных последовательностей утверждений в условном утверждении.

** Прикованные в цепных условиях: **
Условное утверждение с серией альтернативных ветвей.

** Вложенное условное: **
Условное утверждение, которое появляется в одной из ветвей другого условного утверждения.

** рекурсивный: **
Функция, которая вызывает себя, является рекурсивной.

** Базовый случай: **
Условная ветвь в рекурсивной функции, которая не делает рекурсивный вызов.

** Бесконечная рекурсия: **
Рекурсия, у которой нет базового случая или никогда не достигает его.
В конце концов, бесконечная рекурсия вызывает ошибку времени выполнения.

** Newline: **
Персонаж, который создает разрыв линии между двумя частями струны.

### Упражнение

Виртуальные помощники знают о функциях в модуле `jupyturtle ', но есть много версий этих функций, с разными именами, поэтому VA может не знать, о каком вы говорите.

Чтобы решить эту проблему, вы можете предоставить дополнительную информацию, прежде чем задать вопрос.
Например, вы можете начать подсказку с «Вот программа, которая использует модуль` jupyturtle », а затем вставить в один из примеров из этой главы.
После этого VA должен иметь возможность генерировать код, который использует этот модуль.

Например, спросите VA для программы, которая привлекает треугольник Sierpiński.
Код, который вы получаете, должен быть хорошей отправной точкой, но вам, возможно, придется провести некоторую отладку.
Если первая попытка не работает, вы можете рассказать VA, что случилось, и попросить о помощи - или вы можете отладить ее самостоятельно.
