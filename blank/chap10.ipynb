## списки и словари

Вы можете поместить список в словаре в качестве значения.
Например, вот словарь, который отображает от номера «4» в список из четырех букв.


Но вы не можете поместить список в словаре в качестве ключа.
Вот что произойдет, если мы попробуем.


Ранее я упоминал, что словаря используют хэш -таблицы, а это означает, что ключи должны быть ** хэшабл **.

** Хэш ** - это функция, которая принимает значение (любого рода) и возвращает целое число.
Словари используют эти целые числа, называемые хэш -значениями, для хранения и поиска ключей.

Эта система работает только в том случае, если ключ неизменен, поэтому ее хэш -значение всегда одинаковое.
Но если ключ измененен, его хэш -ценность может измениться, и словарь не сработает.
Вот почему клавиши должны быть хранением, и почему изменяемые типы, такие как списки, не являются.

Поскольку словарры изменяются, их также нельзя использовать в качестве ключей.
Но они * могут * использоваться в качестве значений.

## Глоссарий

**словарь:**
 Объект, который содержит пары клавиш, также называемые элементами.

**элемент:**
 В словаре другое имя для пары ключей.

**ключ:**
 Объект, который появляется в словаре как первая часть пары ключей.

**ценить:**
 Объект, который появляется в словаре как вторая часть пары ключевых значений. Это более конкретно, чем наше предыдущее использование слова «значение».

** Картирование: **
  Отношения, в которых каждый элемент из одного набора соответствует элементу другого набора.

** Хэш -таблица: **
Коллекция парных пар, организованных таким образом, чтобы мы могли найти ключ и эффективно найти его значение.

** Hashable: **
 Необываемые типы, такие как целые числа, поплавки и струны, являются хэшабельными.
 Между ними, такими как списки и словари, нет.

** Хэш -функция: **
Функция, которая принимает объект и вычисляет целое число, которое используется для поиска ключа в хэш -таблице.

** Аккумулятор: **
 Переменная, используемая в цикле, чтобы снять или накапливать результат.

** Фильтрация: **
Зацикливание через последовательность и выбор или пропущение элементов.

** Звоните в график: **
Диаграмма, которая показывает каждый кадр, созданный во время выполнения программы, со стрелкой от каждого вызывающего абонента до каждого Callee.

** Памятка: **
 Вычисленное значение, сохраненное, чтобы избежать ненужных будущих вычислений.
