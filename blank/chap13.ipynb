# Файлы и базы данных

Большинство программ, которые мы видели до сих пор, являются**эфемерными**в том смысле, что они запускаются в течение короткого времени и производят выход, но когда они заканчиваются, их данные исчезают.
Каждый раз, когда вы запускаете эфемерную программу, она начинается с чистого сланца.

Другие программы**настойчивы**: они бегают в течение длительного времени (или все время);Они держат хотя бы некоторые из своих данных в долгосрочном хранении;И если они отключены и перезагружаются, они поднимают, где остановились.

Простым способом для сохранения своих данных является чтение и написание текстовых файлов.
Более универсальной альтернативой является хранение данных в базе данных.
Базы данных представляют собой специализированные файлы, которые могут быть прочитаны и написаны более эффективно, чем текстовые файлы, и они предоставляют дополнительные возможности.

В этой главе мы напишем программы, которые читают и пишем текстовые файлы и базы данных, а в качестве упражнения вы напишете программу, которая ищет коллекцию фотографий для дубликатов.
Но прежде чем вы сможете работать с файлом, вы должны его найти, поэтому мы начнем с имен файлов, путей и каталогов.

## имена файлов и пути

Файлы организованы в**каталоги**, также называемые «папки».
Каждая работающая программа имеет**текущий рабочий каталог**, который является каталогом по умолчанию для большинства операций.
Например, когда вы открываете файл, Python ищет его в текущем рабочем каталоге.

Модуль «OS» предоставляет функции для работы с файлами и каталогами («ОС» означает «операционная система»).
Он предоставляет функцию, называемую `getCwd`, которая получает имя текущего рабочего каталога.

Результатом этого примера является домашний каталог пользователя по имени «Dinsdale».
Строка, такая как `'/home/dinsdale'', которая идентифицирует файл или каталог, называется**Path**.

Простое имя файла, подобное «memo.txt'», также считается путем, но это**относительный путь**, потому что оно определяет имя файла относительно текущего каталога.
В этом примере текущий каталог - «/home/dinsdale», поэтому `` memo.txt'` эквивалентен полному пути `'/home/dinsdale/memo.txt'`.

Путь, который начинается с `/`, не зависит от текущего каталога - он называется**абсолютный путь**.
Чтобы найти абсолютный путь к файлу, вы можете использовать `abspath '.

Модуль `OS обеспечивает другие функции для работы с именами файлов и путей.
`listdir` возвращает список содержимого данного каталога, включая файлы и другие каталоги.
Вот пример, который перечисляет содержимое каталога под названием «Фотографии».

Этот каталог содержит текстовый файл с именем `notes.txt` и три каталога.
Катарии содержат файлы изображений в формате JPEG.

Чтобы проверить, существует ли файл или каталог, мы можем использовать `os.path.exists`.

Чтобы проверить, относится ли путь к файлу или каталогу, мы можем использовать `isdir`, который возвращает` true`, если путь относится к каталогу.

И `isfile`, который возвращает` true`, если путь относится к файлу.

Одна из проблем работы с путями заключается в том, что они выглядят по -разному в различных операционных системах.
В системах MacOS и Unix, таких как Linux, имена каталога и файлов в пути разделены прямолинейной чертой, `/`.
Windows использует обратную черту, `\`.
Так что, если вы запускаете эти примеры в Windows, вы увидите обратные черты на пути, и вам придется заменить прямые черты в примерах.

Или, чтобы написать код, который работает в обеих системах, вы можете использовать `os.path.join`, который соединяет каталог и имена файлов в путь, используя прямую или обратную черту, в зависимости от того, какую операционную систему вы используете.

Позже в этой главе мы будем использовать эти функции для поиска набора каталогов и поиск всех файлов изображений.

## отладка

Когда вы читаете и пишете файлы, вы можете столкнуться с проблемами с пробелом.
Эти ошибки могут быть трудно отлаживать, потому что персонажи пробелов обычно невидимы.
Например, вот строка, которая содержит пространства, вкладку, представленную последовательности `\ t`, и новая линия, представленная последовательности` \ n`.
Когда мы печатаем его, мы не видим персонажей пробелов.

Встроенная функция `repr` может помочь.Он принимает любой объект в качестве аргумента и возвращает строковое представление объекта.
Для строк он представляет персонажи пробелов с последовательностями BackSlash.

Это может быть полезно для отладки.

Еще одна проблема, с которой вы можете столкнуться, это то, что разные системы используют разные символы, чтобы указать конец линии.Некоторые системы используют новую линию, представленную `\ n`.Другие используют возвращающийся символ, представленного `\ r`.
Некоторые используют оба.Если вы перемещаете файлы между различными системами, это
Несоответствия могут вызвать проблемы.

Имя файла Капитализация - это еще одна проблема, с которой вы можете столкнуться, если вы работаете с различными операционными системами.
В MacOS и Unix имена файлов могут содержать строчные и верхние буквы, цифры и большинство символов.
Но многие приложения Windows игнорируют разницу между строчными и прописными буквами, и в Windows не допускаются несколько символов, которые разрешены в MacOS и UNIX.

## Глоссарий

**эфемерный:**
Эфемерная программа обычно работает в течение короткого времени, и, когда она заканчивается, ее данные теряются.

**настойчивый:**
Постоянная программа работает на неопределенный срок и хранит, по крайней мере, некоторые из своих данных в постоянном хранении.

**Справочник:**
Сборник файлов и других каталогов.

**Текущий рабочий каталог:**
Каталог по умолчанию, используемый программой, если не указан другой каталог.

**путь:**
Строка, которая указывает последовательность каталогов, часто приводящих к файлу.

**относительный путь:**
Путь, который начинается с текущего рабочего каталога или какого -либо другого указанного каталога.

**Абсолютный путь:**
Путь, который не зависит от текущего каталога.

**F-String:**
Строка, которая имеет букву «f» перед открытием кавычки, и содержит одно или несколько выражений в кудрявых скобках.

**Данные конфигурации:**
Данные, часто хранящиеся в файле, которые указывают, что должна делать программа и как.

**Сериализация:**
Преобразование объекта в строку.

**десериализация:**
Преобразование строки в объект.

**База данных:**
Файл, содержимое которого организовано для эффективного выполнения определенных операций.

**Магазины ключей:**
База данных, содержимое которого организовано как словарь с ключами, которые соответствуют значениям.

**двоичный режим:**
Способ открыть файл, чтобы содержимое интерпретируется как последовательность байтов, а не последовательность символов.

**Хэш -функция:**
Функция, которая берет и объект и вычисляет целое число, которое иногда называют дайджестом.

**Дайджест:**
Результат хэш -функции, особенно когда она используется для проверки, являются ли два объекта одинаковыми.

### Упражнение

В большой коллекции файлов может быть более одной копии одного и того же файла, хранящейся в разных каталогах или с разными именами файлов.
Цель этого упражнения - искать дубликаты.
В качестве примера мы будем работать с файлами изображений в каталоге «Фотографии».

Вот как это будет работать:

*Мы будем использовать функцию `walk` от [](section_walking_directories) для поиска этого каталога для файлов, которые заканчиваются одним из расширений в` config ['agensions'] `.

*Для каждого файла мы используем `md5_digest` from [](section_md5_digest) для вычисления дайджеста содержимого.

*Используя полку, мы сделаем отображение из каждого дайджеста в список путей с этим дайджестом.

*Наконец, мы будем искать полку для любых дайджестов, которые карту в нескольких файлах.

*Если мы найдем что -либо, мы будем использовать `some_contents`, чтобы подтвердить, что файлы содержат те же данные.

Я предложу несколько функций, чтобы написать сначала, затем мы объединим все это.

1. Чтобы идентифицировать файлы изображений, напишите функцию с именем `is_image`, которая берет путь и список расширений файлов, и возвращает` true`, если путь заканчивается одним из расширений в списке.Подсказка: используйте `os.path.splitext` - или попросите виртуального помощника написать эту функцию для вас.

Вы можете использовать `doctest`, чтобы проверить вашу функцию.

2. Напишите функцию с именем `add_path`, которая принимает в качестве аргументов путь и полку.Он должен использовать `md5_digest` для вычисления дадеста содержимого файла.Затем он должен обновить полку, создавая новый элемент, который карты из дайджеста в список, содержащий путь, либо добавление пути к списку, если он существует.

3. Напишите версию «Walk» под названием `walk_images`, которая принимает каталог и проходит через файлы в каталоге и его подкаталогах.Для каждого файла он должен использовать `is_image`, чтобы проверить, является ли это файл изображения и` add_path`, чтобы добавить его на полку.
