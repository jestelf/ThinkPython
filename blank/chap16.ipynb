## Глубокая копия

Когда мы используем `copy` для дублирования` box1`, он копирует объект `rectangle ', но не объект` point', который он содержит.
Таким образом, `box1` и` box2` - это разные объекты, как предполагалось.


Но их атрибуты «угол» относятся к тому же объекту.

То, что делает Copy`, называется ** неглубокая копия **, потому что он копирует объект, но не объекты, которые он содержит.
В результате изменение «ширины» или «высота» одного «прямоугольника» не влияет на другую, но изменение атрибутов общей точки «точка» влияет на оба!
Такое поведение сбивает с толку и подвержено ошибкам.

К счастью, модуль «Copy» предоставляет другую функцию, называемую «DeepCopy», которая копирует не только объект, но и объекты, на которые он ссылается, и объекты * они * ссылаются, и так далее.
Эта операция называется ** глубокая копия **.

## отладка

В этой главе мы столкнулись с тонкой ошибкой, которая произошла, потому что мы создали «точку», которая была разделена двумя объектами «прямоугольника», а затем мы изменили «точку».
В целом, есть два способа избежать таких проблем, как это: вы можете избежать обмена объектами, или вы можете избежать их изменения.

Чтобы избежать обмена объектами, вы можете использовать Deep Copy, как мы делали в этой главе.

Чтобы избежать изменения объектов, рассмотрите возможность замены нечистых функций, таких как `translate` на чистые функции, такие как` transted`.
Например, вот версия `translated`, которая создает новую« точку »и никогда не изменяет свои атрибуты.


Python предоставляет функции, которые облегчают избегать изменения объектов.
Они выходят за рамки этой книги, но если вам любопытно, спросите виртуального помощника: «Как сделать объект Python неизменным?»

Создание нового объекта занимает больше времени, чем изменение существующего, но разница редко имеет значение на практике.
Программы, которые избегают общих объектов и нечистых функций, часто проще в разработке, тестировании и отладке - и лучший вид отладки - это тот вид, который вам не нужно делать.


## Глоссарий

** мелкая копия: **
Операция копирования, которая не копирует вложенные объекты.

** Глубокая копия: **
Операция копирования, которая также копирует вложенные объекты.

**полиморфизм:**
Способность метода или оператора работать с несколькими типами объектов.
