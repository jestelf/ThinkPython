# Функции и интерфейсы

Мы будем использовать этот модуль для написания функций, которые рисуют квадраты, полигоны и круги - и для демонстрации ** дизайна интерфейса **, который является способом разработки функций, которые работают вместе.

## Модуль JupyTurtle

Чтобы использовать модуль `jupyturtle`, мы можем импортировать его так.

`make_turtle` создает ** холст **, который представляет собой пространство на экране, где мы можем нарисовать, и черепаха, которая представлена круговой оболочкой и треугольной головкой.

Расстояние находится в произвольных единицах - фактический размер зависит от экрана вашего компьютера.

Мы будем использовать функции, определенные в модуле `jupyturtle` много раз, поэтому было бы неплохо, если бы нам не пришлось писать имя модуля каждый раз.

По соглашению, Docstrings представляют собой строки с тройным цитированием, также известные как ** многослойные строки **, потому что тройные кавычки позволяют строке охватывать более одной строки.

Докстронг должен:

* Объясните кратко, что делает функция, не вдаваясь в подробности того, как она работает,

* Укажите, какой тип должен быть каждый параметр, если он не очевиден.

Написание такого рода документации является важной частью дизайна интерфейса.
Хорошо разработанный интерфейс должен быть прост в объяснении;Если вам трудно объяснить одну из ваших функций, возможно, интерфейс может быть улучшен.

## отладка

Интерфейс похож на контракт между функцией и вызывающим абонент.А
Абонент соглашается предоставить определенные аргументы, и функция соглашается
Делайте определенную работу.

Например, `polyline` требует трех аргументов: `n` должен быть целым числом;`length` должен быть положительным числом;и `angle` должен быть числом, которое, как понимается, находится в градусах.

Эти требования называются ** предварительные условия **, потому что они должны быть истинными, прежде чем функция начнет выполнять.И наоборот, условия в конце функции - это ** Посткондиционирование **.
Пост -кондиционирование включают предполагаемый эффект функции (например, сегменты линий рисования) и любые побочные эффекты (например, перемещение черепахи или внесение других изменений).

Предварительные условия являются обязанностью звонящего.Если вызывающий абонент нарушает предпосылку, а функция не работает правильно, ошибка находится в вызывающем абоненте, а не на функции.

Если предварительные условия выполняются, а пост -кондиционирования нет, ошибка в функции.Если ваши до- и почтовые ведения ясны, они могут помочь с отладкой.

## Глоссарий

** Дизайн интерфейса: **
Процесс разработки интерфейса функции, который включает в себя параметры, которые он должен взять.

**холст:**
Окно, используемое для отображения графических элементов, включая линии, круги, прямоугольники и другие формы.

** инкапсуляция: **
Процесс преобразования последовательности операторов в определение функции.

**обобщение:**
Процесс замены чего -то излишне конкретного (например, число) на что -то соответствующим образом общее (например, переменная или параметр).

** Аргумент ключевого слова: **
Аргумент, который включает имя параметра.

** Рефакторинг: **
Процесс изменения рабочей программы для улучшения функциональных интерфейсов и других качеств кода.

** План развития: **
Процесс написания программ.

** Докстронг: **
Строка, которая появляется в верхней части определения функции, для документирования интерфейса функции.

** Многослойная строка: **
Строка, заключенная в тройные кавычки, которая может охватывать более одной строки программы.

** Предварительное условие: **
Требование, которое должно быть удовлетворено вызывающим абонент до начала функции.

** Посткондиционирование: **
Требование, которое должно быть удовлетворено функцией, прежде чем она закончится.
