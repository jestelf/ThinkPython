# Списки

В этой главе представлен один из самых полезных встроенных типов Python - списки.
Вы также узнаете больше об объектах и о том, что может произойти, когда несколько переменных ссылаются на один и тот же объект.

## Список - это последовательность

Как и строка, **list** - это последовательность значений. В строке
значения - это символы; в списке они могут быть любого типа.
Значения в списке называются **элементами**.

Существует несколько способов создания нового списка; самый простой - заключить элементы в квадратные скобки (`[` и `]`).
Например, вот список из двух целых чисел.

А вот список из трех струн.

Элементы списка не обязательно должны быть одного типа.
Следующий список содержит строку, поплавок, целое число и даже другой список.

Список в другом списке **вложен**.

Список, не содержащий элементов, называется пустым списком; вы можете создать
с пустыми скобками, `[]`.

Функция `len` возвращает длину списка.

Длина пустого списка - `0`.

## Списки могут быть изменены

Чтобы прочитать элемент списка, мы можем использовать оператор скобки.
Индекс первого элемента - `0`.

В отличие от строк, списки изменчивы. Когда оператор скобки появляется на
левая часть задания, она идентифицирует элемент списка
которые будут назначены.

Второй элемент `numbers`, который раньше был `123`, теперь `17`.

Индексы списка работают так же, как и строковые индексы:

-   В качестве индекса можно использовать любое целочисленное выражение.

-   Если вы попытаетесь прочитать или написать несуществующий элемент, вы получите
    xYZ3ZYX.

-   Если индекс имеет отрицательное значение, он считается в обратном направлении с конца
    списка.

Оператор `in` работает со списками - он проверяет, отображается ли данный элемент в любом месте списка.

Хотя список может содержать другой список, вложенный список по-прежнему считается одним элементом - поэтому в следующем списке есть только четыре элемента.

И `10` не считается элементом `t`, потому что он является элементом вложенного списка, а не `t`.

## Список срезов

Оператор среза работает со списками так же, как он работает со строками.
В следующем примере выбираются второй и третий элементы из списка из четырех букв.

Если вы опустите первый индекс, срез начнется с начала.

Если опустить вторую, срез идет до конца.

Поэтому, если вы опустите оба, срез будет копией всего списка.

Другой способ скопировать список - использовать функцию `list`.

Поскольку `list` - это имя встроенной функции, вам не следует использовать его в качестве имени переменной.

## Алиасинг

Если `a` ссылается на объект, и вы назначаете `b = a`, то обе переменные ссылаются на один и тот же объект.

Диаграмма состояний выглядит так.

Связь переменной с объектом называется * * ссылкой * *.
В этом примере есть две ссылки на один и тот же объект.

Объект с более чем одной ссылкой имеет более одного имени, поэтому мы говорим, что объект имеет **псевдоним**.
Если псевдоним изменен, изменения, внесенные с одним именем, влияют на другое.
В этом примере, если мы изменим объект, на который ссылается `b`, мы также изменим объект, на который ссылается `a`.

Таким образом, мы бы сказали, что `a` «видит» это изменение.
Хотя такое поведение может быть полезным, оно подвержено ошибкам.
В целом, безопаснее избегать псевдонимов при работе с изменяемыми объектами.

Для неизменных объектов, таких как строки, псевдоним не так уж и проблема.

## Отладка

Обратите внимание, что большинство методов списка изменяют аргумент и возвращают `None`.
Это противоположность строковым методам, которые возвращают новую строку и оставляют оригинал в покое.

Если вы привыкли писать строковый код так:

`remove` изменяет список и возвращает `None`, поэтому следующая операция, которую вы выполняете с `t`, вероятно, завершится неудачей.

Это сообщение об ошибке требует некоторого объяснения.
**атрибут** объекта - это переменная или метод, связанный с ним.
В этом случае значение `t` равно `None`, что является объектом `NoneType`, который не имеет атрибута с именем `remove`, поэтому результатом является `AttributeError`.

Если вы видите подобное сообщение об ошибке, вам следует оглянуться назад и посмотреть, не вызвали ли вы метод списка неправильно.

## Глоссарий

**список:**
 Объект, содержащий последовательность значений.

**Element:**
 Одно из значений в списке или другой последовательности.

**вложенный список:**
Список, который является элементом другого списка.

**разделитель:**
 Символ или строка, используемая для указания, где строка должна быть разделена.

**эквивалент:**
 Имеет одинаковое значение.

**идентично:**
 Быть одним и тем же объектом (что подразумевает эквивалентность).

**ссылка:**
 Связь между переменной и ее значением.

**псевдоним:**
Если есть более одногокоторая относится к объекту, объект имеет псевдоним.

**attribute:**
 Одно из именованных значений, связанных с объектом.
