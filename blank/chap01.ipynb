## Струны

В дополнение к числам, Python также может представлять последовательности букв, которые называются струнами, потому что буквы нанизаны вместе, как бусины на ожерелье.

Струны также могут содержать пробелы, пунктуации и цифры.

##Ценности и типы

До сих пор мы видели три типа ценностей:

* '2' является целым числом,

* '42.0' - число с плавающей точкой, и

"Привет" - это строка.

Своеобразная величина называется **типом**.
Каждая ценность имеет тип, или мы иногда говорим, что она принадлежит типу.

Python предоставляет функцию под названием «тип», которая сообщает вам тип любого значения.
Тип целого числа — «int».

Тип числа с плавающей точкой — «плавающий».

Тип струны — «str».

В качестве функций могут использоваться типы «int», «float» и «str».
Например, «int» может взять число с плавающей точкой и преобразовать его в целое число (всегда округляя вниз).

И «плавающий» может преобразовать целое число в значение с плавающей точкой.

## Официальные и естественные языки

Естественные языки - это языки, на которых говорят люди, такие как английский, испанский и французский. Они не были созданы людьми, они развивались естественно.

**Формальные языки - это языки, разработанные людьми для конкретных приложений.
Например, нотация, которую используют математики, является формальным языком, который особенно хорошо обозначает отношения между числами и символами.
Точно так же языки программирования являются формальными языками, которые были разработаны для выражения вычислений.

Хотя формальные и естественные языки имеют некоторые особенности.
Общими являются важные различия:

Естественные языки полны двусмысленности, с которой люди имеют дело.
использование контекстных ключей и другой информации. Формальные языки являются
Он должен быть почти или полностью однозначным, что означает, что
Любая программа имеет одно значение, независимо от контекста.

* Увольнение: чтобы компенсировать двусмысленность и уменьшить недоразумения;
В естественных языках используется избыточность. В результате они являются
Часто глаголы. Формальные языки менее избыточны и более лаконичны.

* Натуральные языки полны идиом и метафор. Формальные языки означают именно то, что они говорят.

Поскольку мы все взрослеем, говоря на естественных языках, иногда трудно приспособиться к формальным языкам.
Формальные языки более плотные, чем естественные, поэтому их чтение занимает больше времени.
Также важна структура, поэтому не всегда лучше читать сверху вниз, слева направо.
Наконец, важны детали. Небольшие ошибки в орфографии и
пунктуация, с которой можно обойтись на естественных языках, может
Большая разница в формальном языке.

Таким образом, вы узнаете сообщения об ошибках, и когда вы увидите их снова, вы узнаете, что они означают.
Лучше совершать ошибки сейчас и сознательно, чем потом и случайно.

1. Вы можете использовать знак минус, чтобы сделать отрицательное число, такое как «-2». Что произойдет, если поставить знак плюс перед номером? Что такое «2++2»?

2. Что произойдет, если у вас есть два значения без оператора между ними, например «4 2»?

3. Если вы называете функцию «круглым» (42.5), что произойдет, если вы оставите одну или обе скобки?
